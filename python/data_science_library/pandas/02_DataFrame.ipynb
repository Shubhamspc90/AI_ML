**3. Create DataFrame**

(A. From Dictionary)


```python
import numpy as np
import pandas as pd 
```

```python
student_data={
    "name": ["Shubham","Dhar","VCP","DSP"],
    "Age": [23,26,29,33],
    "marks":[80,88,94,99]
}
df=pd.DataFrame(student_data)
print(df)
```

**B. Create DataFrame from List**

```python
data=[
    ["shubham",20,90],
    ["Dhar",25,80],
    ["VCP",29,91],
    ["DSP",33,98]
]
df= pd.DataFrame(data,columns=["Name","Age","Marks"])
print(df)
```

**C. Create DataFrame from List of Dictionaries**

```python
data=[
    {"name":"Shubham","Age":20,"Marks":90},
    {"name":"Dhar","Age":25,"Marks":80},
    {"name":"VCP","Age":29,"Marks":88},
    {"name":"DSP","Age":33,"Marks":78}
]
df=pd.DataFrame(data)
print(df)
```

**D. Create DataFrame from NumPy Array**

```python
arr=np.array([
    [1,"shubham",90],
    [2,"Dhar",92],
    [3,"VCP",95],
    [4,"DSP",99]
])
df=pd.DataFrame(
    arr,
    columns=["ID","Name","Marks"])

print(df)
```

**4. View Data**

```python
print(df.head())  # First 5 rows

print()

print(df.tail())  # Last 5 rows
```

**5. DataFrame Attributes**

```python
print(df.shape)

print(df.columns)

print(df.index)

print(df.dtypes)

print(df.size)

# | Function | Meaning        |
# | -------- | -------------- |
# | shape    | Rows × Columns |
# | columns  | Column Names   |
# | index    | Row Numbers    |
# | dtypes   | Data Types     |
# | size     | Total Values   |
```

**6. DataFrame Methods**

```python
print(df.info())

print()

print(df.describe())

# info() Batata hai
# Rows
# Columns
# Missing Values
# Data Types

# describe() Batata hai
# Mean
# Max
# Min
# Std
# Count
```

**7. Selecting Columns**

```python
print(df["Name"])

print()

print(df["Marks"])

print()

print(df[["Name","Marks"]])
```

**8. Selecting Rows**

```python
print(df.loc[0])
print()
print(df.loc[1])
```

**9. Add New Column**

```python
df["City"] = [
    "Delhi",
    "Mumbai",
    "Pune",
    "Vns"
]
print(df)

print()
df["state"] = [
    "UP",
    "UP",
    "UP",
    "UP"
]

print(df)
```

**10. Delete Column**

```python
df = df.drop(
    columns=["City"]
)

print(df)
```

**11. Rename Columns**

```python
df.rename(
    columns={
        "Marks":"Score"
    },
    inplace=True  # Original DataFrame modify hota hai.
)

print(df)
```

**12. Real World Example**

```python
students = {
    "Name":["Rahul","Aman","Riya","Karan"],
    "Age":[20,21,22,20],
    "Marks":[90,85,95,88]
}

df = pd.DataFrame(students)

print("Average Marks :",df["Marks"].mean())

print("Highest :",df["Marks"].max())

print("Lowest :",df["Marks"].min())
```

```python
# # Final Notes
# | Function         | Purpose                  |
# | ---------------- | ------------------------ |
# | `pd.DataFrame()` | Create DataFrame         |
# | `head()`         | First Rows               |
# | `tail()`         | Last Rows                |
# | `info()`         | Dataset Information      |
# | `describe()`     | Statistics               |
# | `shape`          | Rows × Columns           |
# | `columns`        | Column Names             |
# | `dtypes`         | Data Types               |
# | `loc[]`          | Label-based Selection    |
# | `iloc[]`         | Position-based Selection |
# | `drop()`         | Delete Column            |
# | `rename()`       | Rename Column            |

```
