**3. Import Pandas**

```python
import pandas as pd
```

**4. Creating a Series**

```python
marks= pd .Series([1,2,3,4,5,6])  # series = index + values
print(marks)
```

**5. Series from List**

```python
student = ["Rahul", "Aman", "Riya", "Karan"]

series = pd.Series(student)

print(series)
```

**6. Series from Tuple**

```python
numbers = (10,20,30,40)
series= pd.Series(numbers)
print(series)
```

**7. Series from Dictionary**

```python
student={
    "name":"shubham",
    "age":12,
    "salary": 3400.5
}
series=pd.Series(student)
print(series)
```

**8. Custom Index**

```python
marks = pd.Series(
    [85, 90, 88],
    index=["Rahul", "Aman", "Riya"]
)

print(marks)
```

**9. Accessing Elements**

```python
marks = pd.Series(
    [85, 90, 88],
    index=["Rahul", "Aman", "Riya"]
)
print(marks["Rahul"])

print(marks["Riya"])
```

**10. Series Attributes**

```python
marks = pd.Series(
    [85, 90, 88],
    index=["Rahul", "Aman", "Riya"]
)
print(marks.index)
print(marks.values)
print(marks.dtype)
print(marks.shape)
print(marks.size)
```

**11. Basic Operations**

```python
marks = pd.Series(
    [85, 90, 88],
    index=["Rahul", "Aman", "Riya"]
)

print(marks + 5)

print(marks * 2)

print(marks.mean())

print(marks.max())

print(marks.min())
```

**12. Real World Example**

```python
students = pd.Series(
    [78, 85, 92, 67, 88],
    index=["Rahul", "Aman", "Riya", "Karan", "Priya"]
)

print("Average Marks :", students.mean())

print("Highest Marks :", students.max())

print("Lowest Marks :", students.min())
```
