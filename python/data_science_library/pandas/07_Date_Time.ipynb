**2. Current Date & Time**

```python
import pandas as pd
print(pd.Timestamp.now())
```

**3. Convert String to Datetime**

```python
date = pd.to_datetime("2026-01-15")
print(date)

print()

# Multiple Dates
dates = pd.to_datetime([
    "2026-01-01",
    "2026-02-10",
    "2026-03-20"
])
print(dates)
```

**4. Date Attributes**

```python
date= pd.to_datetime("2026/07/15")
print(date.year)
print(date.month)
print(date.day)
print(date.day_name())
print(date.month_name())
```

**5. Date Formatting**

```python
date = pd.Timestamp.now()

print(date.strftime("%d-%m-%Y"))

print()

print(date.strftime("%d/%m/%Y"))

print(date.strftime("%A"))

print(date.strftime("%B"))
```

**6. Extract Date Components**

```python
orders = pd.DataFrame({

    "Order_Date":[

        "2026-01-01",

        "2026-03-15",

        "2026-05-20"

    ]

})

orders["Order_Date"] = pd.to_datetime(
    orders["Order_Date"]
)

orders["Year"] = orders["Order_Date"].dt.year

orders["Month"] = orders["Order_Date"].dt.month

orders["Day"] = orders["Order_Date"].dt.day

print(orders)
```

**7. Extract Hour, Minute, Second**

```python
time = pd.Timestamp.now()
print(time.hour)
print(time.minute)
print(time.second)
print(time.microsecond)
```

**8. Date Arithmetic**

```python
date = pd.to_datetime("2026-01-01")

print(date + pd.Timedelta(days=10))

print(date - pd.Timedelta(days=5))
```

**9. Timedelta**

```python
start = pd.to_datetime("2026-01-01")

end = pd.to_datetime("2026-01-20")

difference = end - start

print(difference)
```

**10. Date Range**

```python
dates = pd.date_range(

    start="2026-01-01",

    end="2026-01-10"

)

print(dates)
```

**11. Date Filtering**

```python
sales = pd.DataFrame({
    
    "Date":[ "2026-01-01", "2026-02-15", "2026-03-20"],
    "Sales":[ 5000, 7000, 9000 ]
})

sales["Date"] = pd.to_datetime( sales["Date"] )

print( sales[ sales["Date"] > "2026-02-01"] )
```

**12. Real World Example**

```python
orders = pd.DataFrame({
    "Order_Date":[ "2026-01-10", "2026-02-15", "2026-03-25"],

    "Amount":[500,900,1200]

})

orders["Order_Date"] = pd.to_datetime(
    orders["Order_Date"]
)

orders["Month"] = orders["Order_Date"].dt.month_name()

print(orders)
```
