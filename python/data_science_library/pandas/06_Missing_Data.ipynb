**1. What is Missing Data?**

```python
# Missing Data ka matlab hai ki dataset me kuch values available nahi hain.
# Pandas me missing values ko NaN (Not a Number) se represent kiya jata hai.
```

**2. Create Sample DataFrame**

```python

import pandas as pd
student = {
        "name":["shubham","Dhar","VCP","DSP","Priya"],
        "age": [20,None,25,29,None],
        "marks":[90,89,None,92,95],
        "city":["Delhi","Mumbai","Delhi",None,"Pune"]    
    }

df = pd.DataFrame(student)
print(df)

```

**3. isna()**

```python
print(df.isna())
# Har cell ke liye
# True → Missing
# False → Available
```

**4. Count Missing Values**

```python

print(df.isna().sum())
```

**5. notna()**

```python
print(df.notna())
# Opposite of isna().
# True → Value Available
# False → Missing
```

**6. dropna()**

```python
print(df.dropna)
# Jis row me ek bhi missing value hai,
# wo row delete ho jayegi.
```

```python
# Sirf poori tarah empty rows delete karna
print(df.dropna(how="all"))
```

```python
# Specific column check karna
print(df.dropna(subset=["marks"]))
```

```python
print(df.dropna(subset=["marks"]))

# 1. dropna()
# dropna() DataFrame se NaN (missing values) ko remove karta hai.

# 2. subset=["marks"]
# Iska matlab:
# Sirf marks column ko check karo.
# Jis row me marks ki value NaN hogi, woh row remove ho jayegi.
```

**7. fillna()**

```python
print(df.fillna(0))
# Saari missing values
# ↓
# 0 se replace ho gayi.

print()
print(df.fillna("Unknown"))
# Saari missing values
# ↓
# Unknown se replace ho gayi.
```

**8. Forward Fill (ffill)**

```python
print(df.ffill())
# Previous valid value ko copy karta hai.
# Example
# 20
# NaN
# 22
# ↓
# 20
# 20
# 22
```

**9. Backward Fill (bfill)**

```python
print(df.bfill())
# Next valid value ko copy karta hai.
# Example
# 20
# NaN
# 22
# ↓
# 20
# 22
# 22
```

**10. Replace with Mean**

```python
df["age"] = df["age"].fillna(
    df["age"].mean()
)

print(df)

# Explanation
# Mean
# ↓
# Average
# Age
# ↓
# Missing values replace.
```

**11. Replace with Median**

```python
df["marks"] = df["marks"].fillna(
    df["marks"].median()
)

print(df)

# Explanation
# Median
# ↓
# Middle Value
# Agar outliers ho
# ↓
# Median better hota hai.
```

**12. Replace with Mode**

```python
df["city"] = df["city"].fillna(
    df["city"].mode()[0]
)

print(df)
# Explanation
# Mode
# ↓
# Most Frequent Value
# City column me mostly string hota hai,
# isliye mode use karte hain.
```

**13. Missing Value Percentage**

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Amit", "Neha", "Rahul", "Priya", "Rohit"],
    "Age": [22, np.nan, 25, np.nan, 28],
    "Marks": [85, 90, np.nan, 75, 88],
    "City": ["Delhi", np.nan, "Noida", "Delhi", "Ghaziabad"]
})

print("DataFrame")
print(df)

print("\nMissing Values")
print(df.isna().sum())

print("\nMissing Value Percentage")
print((df.isna().sum() / len(df)) * 100)
```
