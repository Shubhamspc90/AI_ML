**1. What is merge()?**

```python
# Theory

# merge() ka use do ya do se zyada DataFrames ko common column ke basis par combine karne ke liye hota hai.

# Database me is concept ko JOIN kehte hain.
```

**2. Create Sample DataFrames**

```python
import pandas as pd
student= pd.DataFrame(
    {
        "student_ID":[10,20,30,40],
        "Name":["Shubham","Aman","Riya","Priya"]
    }
)

marks=pd.DataFrame(
    {
        "student_ID":[10,20,30,50],
        "marks":[90,85,95,80]
    }
)

print("Student DataFrame")
print(student)

print()

print("Marks DataFrame")
print(marks)
```

**3. Inner Join**

```python
result = pd.merge(
    student,
    marks,
    on="student_ID"
)

print(result)
```

**4. Left Join**

```python
result =pd.merge(
    student,
    marks,
    on="student_ID",
    how="left"
)
print(result)
```

**5. Right Join**

```python
resulr= pd.merge(
    student,
    marks,
    on="student_ID",
    how="right"
)
print(result)
```

**6. Outer Join**

```python
result=pd.merge(
    student,
    marks,on="student_ID",
    how="outer"
)
print(result)
```

**7. Merge on Different Column Names**

```python
students = pd.DataFrame({
    "ID":[101,102,103],
    "Name":["Shubham","Dhar","VCP"]
})

marks = pd.DataFrame({
    "Student_ID":[101,102,103],
    "Marks":[90,85,95]
})

result = pd.merge(
    students,
    marks,
    left_on="ID",
    right_on="Student_ID"
)

print(result)
```

**8. Merge Multiple DataFrames**

```python
courses = pd.DataFrame({
    "Student_ID":[101,102,103],
    "Course":[
        "Python",
        "Java",
        "Data Science"
    ]
})

result = pd.merge(
    students,
    marks,
    left_on="ID",
    right_on="Student_ID"
)

result = pd.merge(
    result,
    courses,
    on="Student_ID"
)

print(result)
```

**9. join()**

```python
df1 = students.set_index("ID")
df2 = marks.set_index("Student_ID")

print(
    df1.join(df2)
)
```

**10. concat()**

```python
# concat() ka use DataFrames ko upar-niche ya side-by-side jodne ke liye hota hai.
# (Rows)
df1 = pd.DataFrame({
    "Name":["Rahul","Aman"]
})

df2 = pd.DataFrame({
    "Name":["Riya","Priya"]
})

print(
    pd.concat([df1,df2])
)

print()
# (Columns)
name = pd.DataFrame({
    "Name":["Rahul","Aman"]
})

marks = pd.DataFrame({
    "Marks":[90,85]
})

print(
    pd.concat(
        [name,marks],
        axis=1
    )
)
```

**11. Real World Example**

```python
student = pd.DataFrame({
    "Student_ID":[1,2,3],
    "Name":["Rahul","Aman","Riya"]
})

fees = pd.DataFrame({
    "Student_ID":[1,2,3],
    "Fees":[25000,30000,28000]
})

attendance = pd.DataFrame({
    "Student_ID":[1,2,3],
    "Attendance":[90,95,88]
})

result = pd.merge(student, fees, on="Student_ID")
result = pd.merge(result, attendance, on="Student_ID")

print(result)
```
