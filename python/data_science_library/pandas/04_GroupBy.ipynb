**1. What is groupby()?**


```python
# Theory:
# groupby() ka use similar data ko ek group me divide karne ke liye hota hai.
# Uske baad har group par
# Sum
# Mean
# Count
# Maximum
# Minimum
# jaise calculations kar sakte hain.
```

**2. Why groupby()?**

```python

# Theory
# Without groupby
# ↓
# Bahut loops likhne padenge.
# With groupby
# ↓
# Ek line me analysis.
```

**3. Create Sample DataFrame**

```python
import pandas as pd

employees = {
    "Name":["Shubham","Aman","Riya","Priya","Karan","Neha","Ankit"],
    "Department":["IT", "HR","IT","Sales","HR","Sales","IT"],
    "Salary":[50000,40000,65000,45000,42000,55000, 70000]
}

df = pd.DataFrame(employees)

print(df)
```

**4. Basic groupby()**

```python
print(
    df.groupby("Department")
)

# Explanation

# Yeh output directly data nahi dikhata.
# Ye ek GroupBy Object return karta hai.
# Is object ke upar hum calculations karte hain.
```

**5. Sum**

```python
print(
    df.groupby("Department")["Salary"].sum()
)
```

**6. Mean**

```python
print(
    df.groupby("Department")["Salary"].mean()
)
```

**7. Count**

```python
print(
    df.groupby("Department")["Salary"].count()
)
```

**8. Maximum**

```python
print(
    df.groupby("Department")["Salary"].max()
)
```

**9. Minimum**

```python
print(
    df.groupby("Department")["Salary"].min()
)
```

**10. Multiple Aggregations**

```python
print(

    df.groupby("Department")["Salary"]

      .agg(  #ek hi baar me multiple calculations karta hai.

        ["sum","mean","max","min","count"]

      )

)
```

**11. Group By Multiple Columns**

```python

employees["City"] = ["Delhi","Delhi","Mumbai","Delhi","Mumbai","Delhi","Mumbai"]
df = pd.DataFrame(employees)

print(
    df.groupby(
        ["Department","City"]
    )["Salary"].sum()
)
```

**12. agg()**

```python
print(
    df.groupby("Department").agg(
        {
            "Salary":["sum","mean"],
            "Name":"count"
        }
    )
)
```

**13. reset_index()**

```python
result = (

    df.groupby("Department")["Salary"]

      .sum()

      .reset_index()

)

print(result)
```

**14. Real World Example**

```python
print(

    df.groupby("Department")["Salary"]

      .agg(

        ["sum","mean","max","min","count"]

      )

)
```
