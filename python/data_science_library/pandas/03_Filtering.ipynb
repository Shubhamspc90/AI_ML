**1. What is Filtering?**
Filtering ka matlab hai:

Sirf wahi rows nikalna jo kisi condition ko satisfy karti hain.

Example

Agar 5000 students hain aur tumhe sirf


**2. Create Sample DataFrame**

```python
import pandas as pd
students = {
    "Name": ["Shubham", "Aman", "Riya", "Karan", "Priya", "Neha"],
    "Age": [20, 21, 22, 20, 23, 21],
    "Marks": [95, 60, 89, 45, 98, 75],
    "City": ["Delhi", "Mumbai", "Delhi", "Pune", "Delhi", "Mumbai"]
}

df = pd.DataFrame(students)

print(df)
```

**3. Single Condition Filtering**

```python
print(df[df["marks">80]])
```

**4. Multiple Conditions (AND)**

```python
print(
    df[
        ( df["Marks"]>80 ) &
        ( df["City"]=="Delhi")
    ]
)
```

**5. Multiple Conditions (OR)**

```python
print(
    df[
       (df["Marks"]>70) |
       (df["City"]=="Delhi")    
    ]
)
```

**6. NOT Condition**

```python
print(
    df[
        ~(df["City"]=="Delhi")   
    ]
)
```

**7. isin()**

```python
print(
    df[
        df["City"].isin(
            ["Delhi","Mumbai"]
        )
    ]
)
```

**8. between()**

```python
print(
    df[
        df["Marks"].between(60,90)
    ]
)
```

**9. query()**

```python
print(
    df.query("Marks>80")
)
```

**10. sort_values()**

```python
print(
    df.sort_values(
        by = "Marks"
    )
)
```

**11. sort_index()**

```python
print(
    df.sort_index(
        ascending=False
    )
)
```

**12. value_counts()**

```python
print(
    df["City"].value_counts()
)
```

**13. unique()**

```python
print(
    df["City"].unique()
)
```

**14. nunique()**

```python
print(
    df["City"].nunique()
)
```

**15. Real World Example**

```python
print("Students Scoring Above 80")

print(
    df[
        df["Marks"] > 80
    ]
)

print()

print("Students from Delhi")

print(
    df[
        df["City"] == "Delhi"
    ]
)

print()

print("Topper")

print(
    df.sort_values(
        by="Marks",
        ascending=False
    ).head(1)
)
```
