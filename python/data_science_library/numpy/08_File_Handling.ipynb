**1. Save Array (.npy)**

```python
import numpy as np
```

```python

import numpy as np
arr = np.array([10, 20, 30, 40, 50])

np.save("student_marks.npy", arr)

print("Array Saved Successfully")
```

**2. Load Array (.npy)**

```python
loaded_arry=np.load("student_marks.npy")
print(loaded_arry)
```

**3. Save Multiple Arrays (.npz)**

```python
marks = np.array([80, 75, 90])

ages = np.array([21, 22, 20])

np.savez("students.npz", marks=marks, ages=ages)

print("Multiple Arrays Saved")
```

**4. Load Multiple Arrays (.npz)**

```python
data =np.load("students.npz")
print(data["marks"])
print(data["ages"])
```

**5. Save Array to CSV**

```python
arr = np.array([
    [10,20,30],
    [40,50,60]
])

np.savetxt("marks.csv", arr, delimiter="," , fmt="%d")
print("CSV File Saved")

# Notes:
# CSV
# ↓
# Excel
# ↓
# Pandas
# ↓
# Data Analysis
# Sab isi format ko use karte hain.
```

**6. Load CSV File**

```python
data = np.loadtxt("marks.csv", delimiter=",")
print(data)
```

**7. Save Text File**

```python
arr = np.array([100,200,300,400])

np.savetxt("numbers.txt", arr)

print("Text File Saved")
```

**8. Load Text File**

```python
data = np.loadtxt("numbers.txt")
print(data)
```

**9. Complete Practice Program**

```python
# Create Array
arr = np.array([10,20,30,40,50])

# Save
np.save("demo.npy", arr)

# Load
data = np.load("demo.npy")

print("Loaded Array")

print(data)

print()

# CSV

np.savetxt("demo.csv", arr, delimiter=",", fmt="%d")

csv_data = np.loadtxt("demo.csv", delimiter=",")

print("CSV Data")

print(csv_data)

print()

# Multiple Arrays

a = np.array([1,2,3])

b = np.array([4,5,6])

np.savez("demo.npz", first=a, second=b)

files = np.load("demo.npz")

print(files["first"])

print(files["second"])
```
