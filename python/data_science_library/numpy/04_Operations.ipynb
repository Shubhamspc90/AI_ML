**1. Arithmetic Operations**

```python
import numpy as np

arr1 = np.array([10, 20, 30])

arr2 = np.array([1, 2, 3])

print("Addition")
print(arr1 + arr2)

print()

print("Subtraction")
print(arr1 - arr2)

print()

print("Multiplication")
print(arr1 * arr2)

print()

print("Division")
print(arr1 / arr2)

print()

print("Floor Division")
print(arr1 // arr2)

print()

print("Modulus")
print(arr1 % arr2)

print()

print("Power")
print(arr1 ** arr2)
```

**2. Comparison Operations**

```python
arr = np.array([10,20,30,40,50])

print(arr>25)

print(arr<25)

print(arr==30)

print(arr!=30)

print(arr>=30)

print(arr<=30)
```

**3. Universal Functions (ufunc)**

```python
arr=np.array([1,4,9,16,25])

print(np.sqrt(arr))

print(np.square(arr))

print(np.abs([-10,20,-30]))

print(np.power(arr,2))
```

**4. Mathematical Functions**

```python
arr=np.array([10,20,30,40])

print(np.min(arr))

print(np.max(arr))

print(np.sum(arr))

print(np.prod(arr)) #10×20×30×40
```

**5. Statistical Functions**

```python
arr=np.array([10,20,30,40,50])

print(np.mean(arr))  # Average

print(np.median(arr))  # Middle Value

print(np.std(arr))  # Data kitna spread hai.

print(np.var(arr))  # Standard deviation ka square.
```

6. Aggregate Functions

```python
arr=np.array([
[10,20,30],
[40,50,60]
])

print(arr.sum())
print(arr.sum(axis=0))  # Column Wise
print(arr.sum(axis=1))  # Row Wise
```

**7. Broadcasting**

```python
arr=np.array([10,20,30])

print(arr+5)

print(arr*2)
```

**8. Vectorization**  (Vectorization = without loop fast computation)

```python
arr=np.array([1,2,3,4,5])

result=arr*10

print(result)
```
