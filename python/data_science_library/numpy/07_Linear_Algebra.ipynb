**1. Dot Product**

```python
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result= np.dot(arr1,arr2)
print(result)

# Dot Product = Multiplication + Addition
# ML Algorithms me bahut use hota hai.
```

**2. Matrix Multiplication**

```python


A= np.array([
    [1,2],
    [3,4]
])

B=np.array([
    [5,6],
    [7,8]
])

print(A@B)  # @ is  Matrix Multiplication Operator

```

**3. matmul()**

```python
result = np.matmul(A,B)  # Same kaam karta hai A@B jaise
print(result)
```

**4. Transpose**

```python
A= np.array([
    [1,2],
    [3,4]
])
print(A.T)
```

**5. Determinant**

```python
A= np.array([
    [1,2],
    [3,4]
])
det = np.linalg.det(A)  
print(det)

# ad-bc  formula
```

**6. Inverse Matrix**

```python
A= np.array([
    [1,2],
    [3,4]
])
inverse = np.linalg.inv(A)  #  A⁻¹
print(inverse)

# Notes
# Inverse sirf
# Square Matrix
# AND
# Determinant ≠ 0
# par hi possible hai.
```

**7. Trace**

```python
A= np.array([
    [1,2],
    [3,4]
])
print(np.trace(A))  # trace => Diagonal Sum
```

**8. Rank**

```python
A= np.array([
    [1,2],
    [3,4]
])

print(np.linalg.matrix_rank(A))
```

**9. Eigen Values**

```python
A= np.array([
    [1,2],
    [3,4]
])

values,vectors=np.linalg.eig(A)
print(values)
```

**10. Eigen Vectors**

```python
A= np.array([
    [1,2],
    [3,4]
])

print(vectors)
```

**11. Complete Practice Program**

```python
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

print("Dot Product")

print(np.dot(np.array([1,2]),np.array([3,4])))

print()

print("Matrix Multiplication")

print(A@B)

print()

print("Transpose")

print(A.T)

print()

print("Determinant")

print(np.linalg.det(A))

print()

print("Inverse")

print(np.linalg.inv(A))

print()

print("Trace")

print(np.trace(A))

print()

print("Rank")

print(np.linalg.matrix_rank(A))

print()

print("Eigen Values")

values,vectors=np.linalg.eig(A)

print(values)

print()

print("Eigen Vectors")

print(vectors)
```
