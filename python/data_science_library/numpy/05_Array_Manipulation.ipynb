**1. reshape()**

```python
import numpy as np
arr = np.array([1,2,3,4,5,6])

new_arr = arr.reshape(2,3)  # 2row,3col

print(new_arr)
```

**2. flatten()**  
2D Array ko, 1D Array me convert karta hai.
Notes:
Copy return karta hai.
Original Array change nahi hota.

```python
arr = np.array([
    [1,2,3],
    [4,5,6]
])

new_arr = arr.flatten()

print(new_arr)
```

**3. ravel()**

```python
arr = np.array([
    [10,20],
    [30,40]
])
new_arr = arr.ravel() # Ye bhi flatten ki tarah 1D Array banata hai.
print(new_arr)
```

**4. transpose()**

```python
arr1 = np.array([
    [10,20],
    [30,40]
])

print(arr1.T)

print()
arr2 = np.array([
    [2,3],
    [4,5]
])

print(arr2.transpose())
```

**5. resize()**

```python
import numpy as np

arr = np.array([1,2,3,4])

arr.resize((2,2))

print(arr)
```

**6. concatenate()**

```python
arr1 = np.array([1,2,3])

arr2 = np.array([4,5,6])

result = np.concatenate((arr1,arr2))

print(result)
```

**7. split()**

```python
import numpy as np

arr1 = np.array([10,20,30,40,50,60])
result1 = np.split(arr1,2)
print(result1)
print()
arr2 = np.array([10,20,30,40,50,60])
result2 = np.split(arr2,2)
print(result2)
```

**8. hstack()**

```python
arr1 = np.array([1,2])

arr2 = np.array([3,4])

print(np.hstack((arr1,arr2)))  # Horizontal Stack
```

**9. vstack()**

```python
arr1 = np.array([1,2])

arr2 = np.array([3,4])

print(np.vstack((arr1,arr2)))  # vertical Stack
```

**10. column_stack()**

```python
arr1 = np.array([1,2,3,4])

arr2 = np.array([5,6,7,8])

print(np.column_stack((arr1,arr2)))  # column Stack
```

**11. Complete Practice Program**

```python
import numpy as np

arr = np.array([1,2,3,4,5,6])

print("Original")
print(arr)

print()

print("Reshape")
print(arr.reshape(2,3))

print()

print("Flatten")
print(arr.reshape(2,3).flatten())

print()

print("Transpose")
print(arr.reshape(2,3).T)

print()

print("Concatenate")

arr2 = np.array([7,8,9])

print(np.concatenate((arr[:3],arr2)))

print()

print("Split")

print(np.split(arr,3))
```
