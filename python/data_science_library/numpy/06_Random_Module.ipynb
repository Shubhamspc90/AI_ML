**1. np.random.rand()**

```python
import numpy as np
arr=np.random.rand(5)
print(arr)
```

**2. np.random.randn()**

```python
arr = np.random.randn(5)

print(arr)
```

**3. np.random.randint()**

```python
arr=np.random.randint(1,100,10)
print(arr)
# Start = 1
# Stop = 100
# Size = 10
```

**4. np.random.random()**

```python
arr = np.random.random(5)

print(arr)
```

**5. np.random.choice()**

```python
arr = np.random.choice([10,20,40,50])

print(arr)
```

**6. np.random.shuffle()**

```python
arr=np.array([10,20,30,40,50])
np.random.shuffle(arr)
print(arr)
```

**7. np.random.permutation()**

```python
arr=np.array([10,20,30,40])
new_arr=np.random.permutation(arr) #Permutation ek shuffled copy return karta hai.
print(new_arr)
print(arr)
```

**8. np.random.seed()**

```python
np.random.seed(100)

print(np.random.randint(1,100,5))
```

**9. Random Sampling**

```python
data = np.array([10,20,30,40,50,60,70])

sample = np.random.choice(data,3)

print(sample)
```

**10. Complete Practice Program**

```python
print("Random Float")

print(np.random.rand(5))

print()

print("Random Integer")

print(np.random.randint(1,100,5))

print()

print("Random Choice")

print(np.random.choice([100,200,300,400]))

print()

arr = np.array([1,2,3,4,5])

print("Shuffle")

np.random.shuffle(arr)

print(arr)

print()

print("Permutation")

print(np.random.permutation(arr))
```
