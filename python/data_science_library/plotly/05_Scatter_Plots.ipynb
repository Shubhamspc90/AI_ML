**Code Cell 1**

```python
# ============================================================
# Import Libraries
# ============================================================

import pandas as pd
import plotly.express as px

df = pd.read_csv("../datasets/raw/sales.csv")

df.head()
```

**Code Cell 2
Basic Scatter Plot**

```python
fig = px.scatter(
    df,
    x="Sales",
    y="Profit",
    title="Sales vs Pofit"
)
fig .show()
```

**Code Cell 3
Scatter with Color**

```python
fig = px.scatter(

    df,

    x="Sales",

    y="Profit",

    color="Category",

    title="Sales vs Profit"

)

fig.show()
```

**Code Cell 4
Bubble Chart**

```python
fig = px.scatter(

    df,

    x="Sales",

    y="Profit",

    color="Category",

    size="Quantity",

    hover_name="Product_Name",

    title="Sales vs Profit"

)

fig.show()
```

**Code Cell 5
Customer Segment Analysis**

```python
fig = px.scatter(

    df,

    x="Discount",

    y="Profit",

    color="Customer_Segment",

    size="Sales",

    title="Discount vs Profit"

)

fig.show()
```

**Code Cell 6
Region Wise Analysis**

```python
fig = px.scatter(

    df,

    x="Sales",

    y="Profit",

    color="Region",

    title="Region Wise Sales"

)

fig.show()
```

**Code Cell 7 (Trendline)**

```python
fig = px.scatter(

    df,

    x="Sales",

    y="Profit",

    trendline="ols",

    title="Sales vs Profit with Trendline"

)

fig.show()
```

**Code Cell 8
Dark Theme**

```python
fig = px.scatter(

    df,

    x="Sales",

    y="Profit",

    color="Category",

    template="plotly_dark"

)

fig.show()
```
