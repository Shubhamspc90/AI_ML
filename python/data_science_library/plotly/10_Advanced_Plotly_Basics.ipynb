**Code Cell 1
3D Scatter Plot**

```python
# Use: ML visualization, clustering, presentations.

import pandas as pd
import plotly.express as px

df = pd.read_csv("../datasets/raw/sales.csv")

fig = px.scatter_3d(
    df,
    x="Sales",
    y="Profit",
    z="Quantity",
    color="Category",
    title="3D Sales Analysis"
)

fig.show()

```

**Code Cell 2
Animation**

```python
# 1. Dataset load
df = pd.read_csv("../datasets/raw/sales.csv")

# 2. Date convert
df["Order_Date"] = pd.to_datetime(
    df["Order_Date"],
    dayfirst=True
)

# 3. Year create
df["Year"] = df["Order_Date"].dt.year

# 4. Verify
print(df.columns)

# 5. Plot
fig = px.scatter(
    df,
    x="Sales",
    y="Profit",
    color="Category",
    animation_frame="Year"
)

fig.show()
```

```python
# Use: Time-based dashboards.

fig = px.scatter(
    df,
    x="Sales",
    y="Profit",
    color="Category",
    animation_frame="Category",
    title="Sales vs Profit by Category"
)

fig.show()
```

```python
# Use: Time-based dashboards.

fig = px.scatter(
    df,
    x="Sales",
    y="Profit",
    color="Payment_Mode",
    animation_frame="Payment_Mode"
)

fig.show()
```

**Code Cell 3
Graph Objects**

```python
# Use: Jab Plotly Express se zyada customization chahiye.

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(

    go.Bar(

        x=["A","B","C"],

        y=[100,200,150]

    )

)

fig.show()
```

**Code Cell 4
Subplots**

```python
# Use: Dashboard me multiple charts ek screen par.

from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(

    rows=1,

    cols=2,

    subplot_titles=("Sales", "Profit")

)

fig.add_trace(

    go.Bar(

        x=["A","B","C"],

        y=[100,200,150]

    ),

    row=1,

    col=1

)

fig.add_trace(

    go.Bar(

        x=["A","B","C"],

        y=[40,70,60]

    ),

    row=1,

    col=2

)

fig.show()
```

**Code Cell 5
Export HTML**

```python
# Interactive chart browser me open ho jayega.

fig.write_html("sales_dashboard.html")
```

Export PNG**

```python
# pip install kaleido

#Presentation aur reports ke liye.
fig.write_image("sales_chart.png")
```
