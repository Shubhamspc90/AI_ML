```python
# # Plotly Box Plots

# ## What is a Box Plot?

# A Box Plot is used to understand the **distribution of numerical data** and to identify **outliers**.

# It summarizes the data using:

# - Minimum
# - First Quartile (Q1)
# - Median (Q2)
# - Third Quartile (Q3)
# - Maximum
# - Outliers

# ## Why do we use Box Plots?

# - Detect Outliers
# - Compare Distributions
# - Understand Spread of Data
# - Compare Categories

# ## Real World Examples

# - Salary Analysis
# - Sales Analysis
# - Student Marks
# - Product Prices
# - Profit Analysis

# ## In this Notebook

# - Basic Box Plot
# - Box Plot by Category
# - Colored Box Plot
# - Horizontal Box Plot
# - Points in Box Plot
# - Business Examples
```

**Code Cell 1**

```python
# ============================================================
# Import Libraries
# ============================================================

import pandas as pd
import plotly.express as px

df = pd.read_csv("../datasets/raw/sales.csv")

df.head()
```

**Code Cell 2
Basic Box Plot**

```python
fig = px.box(

    df,

    y="Profit",

    title="Profit Distribution"

)

fig.show()
```

**Code Cell 3
Category-wise Box Plot**

```python
fig = px.box(

    df,

    x="Category",

    y="Profit",

    title="Profit by Category"

)

fig.show()
```

**Code Cell 4
Colored Box Plot**

```python
fig = px.box(

    df,

    x="Category",

    y="Profit",

    color="Category",

    title="Category-wise Profit"

)

fig.show()
```

**Code Cell 5
Region-wise Box Plot**

```python
fig = px.box(

    df,

    x="Region",

    y="Sales",

    color="Region",

    title="Sales by Region"

)

fig.show()
```

**Code Cell 6
Show All Data Points**

```python
fig = px.box(

    df,

    x="Category",

    y="Profit",

    points="all",

    title="Profit with All Data Points"

)

fig.show()
```

**Code Cell 7
Suspected Outliers**

```python
fig = px.box(

    df,

    x="Category",

    y="Profit",

    points="suspectedoutliers",

    title="Suspected Outliers"

)

fig.show()
```

**Code Cell 8
Horizontal Box Plot**

```python
fig = px.box(

    df,

    x="Profit",

    orientation="h",

    title="Horizontal Profit Distribution"

)

fig.show()
```

**Code Cell 9
Dark Theme**

```python
fig = px.box(

    df,

    x="Category",

    y="Profit",

    color="Category",

    template="plotly_dark"

)

fig.show()
```
