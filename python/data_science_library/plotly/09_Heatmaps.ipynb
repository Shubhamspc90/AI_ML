```python
# # Plotly Heatmaps

# ## What is a Heatmap?

# A Heatmap is a graphical representation of data where values are represented by colors.

# It helps us quickly identify patterns, relationships, and correlations.

# ## Why do we use Heatmaps?

# - Find correlation between numerical columns
# - Detect strong positive/negative relationships
# - Feature Selection for Machine Learning
# - Business Data Analysis

# ## Real World Examples

# - Sales Correlation
# - Student Performance
# - Stock Market Analysis
# - Customer Behavior
# - Machine Learning Feature Selection

# ## What You Will Learn

# - Correlation Matrix
# - Correlation Heatmap
# - Color Scale
# - Text Values
# - Business Example
```

**Code Cell 1**

```python
# ============================================================
# Import Libraries
# ============================================================

import pandas as pd
import plotly.express as px

df = pd.read_csv("../datasets/raw/sales.csv")

df.head()
```

**Code Cell 2
Select Numerical Columns**

```python
numeric_df = df.select_dtypes(include="number")

numeric_df.head()
```

**Code Cell 3
Correlation Matrix**

```python
corr = numeric_df.corr()

corr
```

**Code Cell 4
Basic Heatmap**

```python
fig = px.imshow(
    corr,
    text_auto=True, # values ko cells ke andar show karta hai.
    title="Correlation Heatmap"
)
fig.show()
```

**Code Cell 5
Heatmap with Color Scale**

```python
fig = px.imshow(

    corr,

    text_auto=True,

    color_continuous_scale="RdBu",

    title="Correlation Heatmap"

)

fig.show()
```

**Code Cell 6
Viridis Color Scale**

```python
fig = px.imshow(

    corr,

    color_continuous_scale="Viridis",

    text_auto=True

)

fig.show()
```

**Code Cell 7
Plasma Color Scale**

```python
fig = px.imshow(

    corr,

    color_continuous_scale="Plasma",

    text_auto=True

)

fig.show()
```

**Code Cell 8
Business Example**

```python
important_columns = df[

    [

        "Sales",

        "Profit",

        "Quantity",

        "Discount",

    ]

]

corr = important_columns.corr()

fig = px.imshow(

    corr,

    text_auto=True,

    color_continuous_scale="RdBu",

    title="Business Correlation"

)

fig.show()
```
