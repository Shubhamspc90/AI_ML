**Code Cell 1**

```python
# ============================================================
# Import Libraries
# ============================================================

import pandas as pd
import plotly.express as px

df = pd.read_csv("../datasets/raw/sales.csv")

df.head()
```

**Code Cell 2
Category Wise Sales**

```python
category_sales = (
    df
    .groupby("Category", as_index=False)["Sales"]
    .sum()
)

fig = px.bar(
    category_sales,
    x="Category",
    y="Sales",
    title="Category Wise Sales"
)

fig.show()
```

**Code Cell 3
Horizontal Bar Chart**

```python
fig = px.bar(

    category_sales,

    x="Sales",

    y="Category",

    orientation="h",

    title="Category Wise Sales"

)

fig.show()
```

**Code Cell 4
Color Bar Chart**

```python
fig = px.bar(

    category_sales,

    x="Category",

    y="Sales",

    color="Category",

    title="Category Wise Sales"

)

fig.show()
```

**Code Cell 5
Sales by Region**

```python
region_sales = (

    df

    .groupby("Region", as_index=False)["Sales"]

    .sum()

)

fig = px.bar(

    region_sales,

    x="Region",

    y="Sales",

    color="Region",

    title="Region Wise Sales"

)

fig.show()
```

**Code Cell 6
Top 10 Products**

```python
top_products = (

    df

    .groupby("Product_Name", as_index=False)["Sales"]

    .sum()

    .sort_values("Sales", ascending=False)

    .head(10)

)

fig = px.bar(

    top_products,

    x="Product_Name",

    y="Sales",

    color="Sales",

    title="Top 10 Products"

)

fig.show()
```

**Code Cell 7
Customer Segment**

```python
segment_sales = (

    df

    .groupby("Customer_Segment", as_index=False)["Sales"]

    .sum()

)

fig = px.bar(

    segment_sales,

    x="Customer_Segment",

    y="Sales",

    color="Customer_Segment"

)

fig.show()
```

**Code Cell 8
Profit by Brand**

```python
brand_profit = (

    df

    .groupby("Brand", as_index=False)["Profit"]

    .sum()

)

fig = px.bar(

    brand_profit,

    x="Brand",

    y="Profit",

    color="Profit"

)

fig.show()
```

**Code Cell 9
Text Labels**

```python
fig = px.bar(

    category_sales,

    x="Category",

    y="Sales",

    text="Sales"

)

fig.show()
```

**Code Cell 10
Dark Theme**

```python
fig = px.bar(

    category_sales,

    x="Category",

    y="Sales",

    template="plotly_dark",

    color="Category"

)

fig.show()
```
