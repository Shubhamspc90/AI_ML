```python
# # What is Plotly?

# Plotly is an open-source Python visualization library used to create **interactive graphs and dashboards**.

# Unlike Matplotlib and Seaborn, Plotly graphs are interactive by default.

# You can:

# - Zoom
# - Pan
# - Hover
# - Select data
# - Save graphs
# - Build dashboards

# without writing extra code.
```

**Code Cell 1**

```python
# ============================================================
# Check Plotly Installation
# ============================================================

import plotly

print("Plotly Version :", plotly.__version__)
```

```python
# # Why Plotly?

# Matplotlib creates static charts.

# Seaborn creates beautiful statistical charts.

# Plotly creates **interactive charts**.

# It is widely used in:

# - Business Dashboards
# - Data Analysis
# - Data Science
# - Machine Learning
# - Financial Analysis
# - Power BI alternatives
```

```python
# | Feature            | Matplotlib | Seaborn | Plotly |
# | ------------------ | ---------- | ------- | ------ |
# | Static Charts      | ✅          | ✅       | ❌      |
# | Interactive Charts | ❌          | ❌       | ✅      |
# | Zoom               | ❌          | ❌       | ✅      |
# | Hover Information  | ❌          | ❌       | ✅      |
# | Dashboard Support  | ❌          | ❌       | ✅      |
# | Web Ready          | ❌          | ❌       | ✅      |
```

```python
# # Plotly Modules

# There are two major ways to create charts:

# ## 1. Plotly Express (px)

# Easy

# Less code

# Best for beginners

# ---

# ## 2. Graph Objects (go)

# Advanced

# More customization

# Used in professional dashboards.
```

**Code Cell 2**

```python
# ============================================================
# Import Plotly
# ============================================================

import plotly.express as px

import plotly.graph_objects as go

print("Plotly Imported Successfully")
```

**Code Cell 3**

```python
# # First Interactive Chart
# Let's create our first Plotly chart.

# ============================================================
# Simple Line Chart
# ============================================================

import pandas as pd

df = pd.DataFrame({

    "Month": ["Jan","Feb","Mar","Apr","May"],

    "Sales": [100,120,180,160,210]

})

fig = px.line(

    df,

    x="Month",

    y="Sales",

    title="Monthly Sales"

)

fig.show()
```

**Code Cell 4(Interactive Scatter Plot)**

```python
fig = px.scatter(
    df,
    x="Month",
    y="Sales",
    title="Intractive Scatter Plot"
)
fig.show()
```

**Code Cell 5 ( Interactive Bar Chart )**

```python
fig = px.bar(
    df,
    x="Month",
    y="Sales",
    title="Monthly Sales"
)

fig.show()
```

**Code Cell 6 (Interactive Pie Chart)**

```python
fig = px.pie(
    df,
    names="Month",
    values="Sales",
    title="Sales Share"
)

fig.show()
```
