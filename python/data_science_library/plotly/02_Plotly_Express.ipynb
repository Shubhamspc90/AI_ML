```python
# # What is Plotly Express?

# Plotly Express (px) is the high-level interface of Plotly.

# It allows us to create interactive charts using a single line of code.

# Example:

# fig = px.bar(...)
# fig.show()

# It is recommended for:

# - Beginners
# - Data Analysis
# - Exploratory Data Analysis (EDA)
# - Dashboards
```

**Code Cell 1 (Import Libraries)**

```python
# ============================================================
# Import Libraries
# ============================================================

import pandas as pd
import plotly.express as px

print("Libraries Imported Successfully")
```

**Code Cell 2**
Load Dataset

```python
# ============================================================
# Load Sales Dataset
# ============================================================

df = pd.read_csv("../datasets/raw/sales.csv")

print(df.shape)

df.head()
```

```python
# # Syntax of Plotly Express

# fig = px.chart_type(

#     data_frame,

#     x = "...",

#     y = "...",

#     color = "...",

#     title = "..."

# )

# fig.show()
```

**Code Cell 3**
Simple Line Chart

```python
fig = px.line(
    df,
    x="Order_Date",
    y="Sales",
    title="Daily Sales"
)

fig.show()
```

**Code Cell 4**Line Chart with Color

```python
fig = px.line(
    df,
    x="Order_Date",
    y="Sales",
    color="Category",
    title="Sales by Category"
)

fig.show()
```

**Code Cell 5
Interactive Bar Chart**

```python
category_sales = (
    df
    .groupby("Category",as_index=False)["Sales"]
    .sum()
)

fig = px.bar(
    category_sales,
    x="Category",
    y="Sales",
    title="Category Wise Sales"
)

fig.show()
```

**Code Cell 6
Horizontal Bar Chart**

```python
fig = px.bar(
    category_sales,
    x="Sales",
    y="Category",
    orientation="h",
    title="Horizontal Bar Chart"
)

fig.show()
```

**Code Cell 7
Colored Bar Chart**

```python
fig = px.bar(
    category_sales,
    x="Category",
    y="Sales",
    color="Category",
    title="Category Sales"
)

fig.show()
```

**Code Cell 8
Scatter Plot**

```python
fig = px.scatter(
    df,
    x="Sales",
    y="Profit",
    title="Sales vs Profit"
)

fig.show()
```

**Code Cell 9
Scatter with Color**

```python
fig = px.scatter(
    df,
    x="Sales",
    y="Profit",
    color="Category",
    title="Sales vs Profit"
)

fig.show()
```

**Code Cell 10
Bubble Chart**

```python
fig = px.scatter(
    df,
    x="Sales",
    y="Profit",
    color="Category",
    size="Quantity",
    hover_name="Product_Name",
    title="Bubble Chart"
)

fig.show()
```

**Code Cell 11
Histogram**

```python
fig = px.histogram(
    df,
    x="Sales",
    nbins=40,
    title="Sales Distribution"
)

fig.show()
```

**Code Cell 12
Box Plot**

```python
fig = px.box(
    df,
    y="Profit",
    title="Profit Distribution"
)

fig.show()
```

**Code Cell 13
Violin Plot**

```python
fig = px.violin(

    df,

    y="Profit",

    box=True,

    title="Profit Distribution"

)

fig.show()
```

**Code Cell 14
Area Chart**

```python
monthly = (

    df

    .groupby("Order_Date",as_index=False)["Sales"]

    .sum()

)

fig = px.area(

    monthly,

    x="Order_Date",

    y="Sales",

    title="Area Chart"

)

fig.show()
```

**Code Cell 15
Funnel Chart**

```python
segment = (

    df

    .groupby("Customer_Segment",as_index=False)["Sales"]

    .sum()

)

fig = px.funnel(

    segment,

    x="Sales",

    y="Customer_Segment",

    title="Sales Funnel"

)

fig.show()
```
