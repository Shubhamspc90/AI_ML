```python
# # Plotly Pie Charts

# ## What is a Pie Chart?

# A Pie Chart is used to show how different categories contribute to a whole.

# Each slice represents a category, and its size represents its percentage or value.

# ## Why do we use Pie Charts?

# - Show percentage contribution
# - Compare categories
# - Business dashboards
# - Sales reports

# ## Real World Examples

# - Sales by Category
# - Revenue by Region
# - Market Share
# - Customer Segments
# - Payment Methods

# ## In this Notebook

# - Basic Pie Chart
# - Colored Pie Chart
# - Donut Chart
# - Pie Chart with Percentage
# - Business Examples
```

**Code Cell 1**

```python
import pandas as pd
import plotly.express as px
df = pd.read_csv("../datasets/raw/sales.csv")
df.head()
```

**Code Cell 2
Category Wise Sales**

```python
category_sales = (

    df
    .groupby("Category", as_index=False)["Sales"]
    .sum()

)

fig = px.pie(

    category_sales,

    names="Category",

    values="Sales",

    title="Category Wise Sales"

)

fig.show()
```

**Code Cell 3
Region Wise Sales**

```python
region_sales = (

    df
    .groupby("Region", as_index=False)["Sales"]
    .sum()

)

fig = px.pie(

    region_sales,

    names="Region",

    values="Sales",

    title="Region Wise Sales"

)

fig.show()
```

**Code Cell 4
Donut Chart**

```python
fig = px.pie( # Pie Chart ko Donut Chart bana deta hai.

    category_sales,

    names="Category",

    values="Sales",

    hole=0.5,

    title="Category Wise Sales"

)

fig.show()
```

**Code Cell 5
Customer Segment**

```python
segment_sales = (

    df
    .groupby("Customer_Segment", as_index=False)["Sales"]
    .sum()

)

fig = px.pie(

    segment_sales,

    names="Customer_Segment",

    values="Sales",

    title="Customer Segment Contribution"

)

fig.show()
```

**Code Cell 6
Payment Mode**

```python
payment_sales = (

    df
    .groupby("Payment_Mode", as_index=False)["Sales"]
    .sum()

)

fig = px.pie(

    payment_sales,

    names="Payment_Mode",

    values="Sales",

    title="Payment Mode Analysis"

)

fig.show()
```

**Code Cell 7
Show Percentage**

```python
fig = px.pie(

    category_sales,

    names="Category",

    values="Sales"

)

fig.update_traces(

    textinfo="percent+label"

)

fig.show()
```

**Code Cell 8
Dark Theme**

```python
fig = px.pie(

    category_sales,

    names="Category",

    values="Sales",

    hole=0.4,

    template="plotly_dark"

)

fig.show()
```
