**Code Cell 1 Import Libraries**

```python
# ============================================================
# Import Libraries
# ============================================================

import pandas as pd
import plotly.express as px

df = pd.read_csv("../datasets/raw/sales.csv")

df["Order_Date"] = pd.to_datetime(df["Order_Date"],format="%d-%m-%Y")

print(df.head())
```

**Code Cell 2
Daily Sales**

```python
daily_sales = (
    df
    .groupby("Order_Date", as_index=False)["Sales"]
    .sum()
)

fig = px.line(
    daily_sales,
    x="Order_Date",
    y="Sales",
    title="Daily Sales"
)

fig.show()
```

**Code Cell 3
Monthly Sales**

```python
monthly_sales = (

    df
    .groupby(df["Order_Date"].dt.to_period("M"))["Sales"]
    .sum()
    .reset_index()

)

monthly_sales["Order_Date"] = monthly_sales["Order_Date"].astype(str)

fig = px.line(

    monthly_sales,

    x="Order_Date",

    y="Sales",

    title="Monthly Sales"

)

fig.show()
```

**Code Cell 4
Multiple Line Chart**

```python
category_sales = (

    df
    .groupby(

        ["Order_Date","Category"],

        as_index=False

    )["Sales"]

    .sum()

)

fig = px.line(

    category_sales,

    x="Order_Date",

    y="Sales",

    color="Category",

    title="Category Wise Sales"

)

fig.show()
```

**Code Cell 5
Line With Markers**

```python
fig = px.line(

    monthly_sales,

    x="Order_Date",

    y="Sales",

    markers=True,

    title="Monthly Sales with Markers"

)

fig.show()
```

**Code Cell 6
Change Line Color**

```python
fig = px.line(

    monthly_sales,

    x="Order_Date",

    y="Sales",

    color_discrete_sequence=["red"]

)

fig.show()
```

**Code Cell 7
Different Line Style**

```python
fig = px.line(

    category_sales,

    x="Order_Date",

    y="Sales",

    color="Category",

    line_dash="Category"

)

fig.show()
```

**Code Cell 8
Hover Data**

```python
fig = px.line(

    category_sales,

    x="Order_Date",

    y="Sales",

    color="Category",

    hover_data=["Category","Sales"]

)

fig.show()
```

**Code Cell 9
Range Slider**

```python
fig = px.line(

    daily_sales,

    x="Order_Date",

    y="Sales"

)

fig.update_xaxes(

    rangeslider_visible=True

)

fig.show()
```

**Code Cell 10
Range Selector**

```python
fig = px.line(

    daily_sales,

    x="Order_Date",

    y="Sales"

)

fig.update_xaxes(

    rangeselector=dict(

        buttons=list([

            dict(count=7,label="7D",step="day",stepmode="backward"),

            dict(count=1,label="1M",step="month",stepmode="backward"),

            dict(count=6,label="6M",step="month",stepmode="backward"),

            dict(step="all")

        ])

    )

)

fig.show()
```

**Code Cell 11
Dark Theme**

```python
fig = px.line(

    monthly_sales,

    x="Order_Date",

    y="Sales",

    template="plotly_dark",

    markers=True

)

fig.show()
```

**Code Cell 12
White Theme**

```python
fig = px.line(

    monthly_sales,

    x="Order_Date",

    y="Sales",

    template="plotly_white"

)

fig.show()
```

**Code Cell 13
Business Example**

```python
region_sales = (

    df

    .groupby(

        ["Order_Date","Region"],

        as_index=False

    )["Sales"]

    .sum()

)

fig = px.line(

    region_sales,

    x="Order_Date",

    y="Sales",

    color="Region",

    markers=True,

    title="Region Wise Sales Trend"

)

fig.show()
```
