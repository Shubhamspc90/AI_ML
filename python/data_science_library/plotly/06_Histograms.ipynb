```python
# # Plotly Histograms

# ## What is a Histogram?

# A Histogram is used to visualize the **distribution of numerical data**.

# Unlike a Bar Chart, a Histogram groups continuous values into ranges (called **bins**) and shows how many observations fall into each range.

# ### Why do we use Histograms?

# - Understand data distribution
# - Detect skewness
# - Find the most common value range
# - Identify outliers
# - Prepare data for Machine Learning

# ### Real-world Examples

# - Customer Age Distribution
# - Salary Distribution
# - Product Price Distribution
# - Student Marks Distribution
# - Sales Distribution

# ### In this Notebook

# - Basic Histogram
# - Histogram with Colors
# - Histogram by Category
# - Number of Bins
# - Histogram with Marginal Box Plot
# - Business Examples
```

**Code Cell 1**

```python
# ============================================================
# Import Libraries
# ============================================================

import pandas as pd
import plotly.express as px

df = pd.read_csv("../datasets/raw/sales.csv")

df.head(50)
```

**Code Cell 2
Basic Histogram**

```python
fig = px.histogram(
    df,
    x="Sales",
    title="Sales Distribution"
)
fig.show()
```

**Code Cell 3
Number of Bins**

```python
fig = px.histogram(

    df,

    x="Sales",

    nbins=30,

    title="Sales Distribution"

)

fig.show()
```

**Code Cell 4
Histogram with Color**

```python
fig = px.histogram(

    df,

    x="Sales",

    color="Category",

    title="Sales Distribution by Category"

)

fig.show()
```

**Code Cell 5
Profit Distribution**

```python
fig = px.histogram(

    df,

    x="Profit",

    title="Profit Distribution"

)

fig.show()
```

**Code Cell 6
Customer Rating Distribution**

```python
fig = px.histogram(

    df,

    x="Customer_Rating",

    nbins=10,

    title="Customer Rating Distribution"

)

fig.show()
```

**Code Cell 7
Histogram + Box Plot**

```python
fig = px.histogram(

    df,

    x="Sales",

    marginal="box",
    
    title="Sales Distribution with Box Plot"

)

fig.show()

# marginal="box" histogram ke upar box plot add karta hai, 
# jisse outliers aur spread samajhna aasaan ho jata hai.
```

**Code Cell 8
Histogram + Rug Plot**

```python
fig = px.histogram(

    df,

    x="Sales",

    marginal="rug",

    title="Sales Distribution with Rug Plot"

)

fig.show()
# Rug Plot har individual observation ko chhoti lines ke form me dikhata hai.
```
