```python
# Notes
# pairplot() → Sab numerical columns compare karta hai.
# hue → Different categories ko color deta hai.
# diag_kind="hist" → Histogram.
# diag_kind="kde" → Density Curve.
# corner=True → Half graph show karta hai.
# vars=[] → Selected columns.
# plot_kws → Scatter customization.
# diag_kws → Diagonal customization.
```

**1. What is Pairplot?**

```python
# Pairplot automatically

# Har numerical column ki distribution
# Har numerical column ka relationship
# Scatter Plots
# Histograms/KDE

# ek hi figure me bana deta hai.

# Matlab agar dataset me 4 numerical columns hain to Pairplot automatically un sabka comparison bana dega.

# Real World Example

# Ek Data Scientist ke paas Student Dataset hai.
# Math
# Science
# English
# Computer

# Usse dekhna hai

# Math vs Science
# Math vs English
# Science vs Computer
# English vs Computer

# Agar manually graph banayega to bahut time lagega.
# Pairplot
# ↓
# Ek command me sab graphs.
```

**2. Why Pairplot?**

```python
# Pairplot use hota hai

# Exploratory Data Analysis (EDA)
# Machine Learning
# Feature Selection
# Relationship Detection
# Outlier Detection
# Pattern Recognition
```

```python
# Dataset Load
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
print(iris.head())
```

**3. Basic pairplot()**

```python
# Ye automatically
# Scatter Plots
# Histograms
# create karta hai.

sns.pairplot(iris)
plt.show()
```

**4. hue Parameter**

```python
sns.pairplot(
    iris,
    hue="species"
)
plt.show()
```

**5. diag_kind**

```python
# Diagonal me kya show hoga.
# Default
# ↓
# Histogram


# Histogram
sns.pairplot(
    iris,
    diag_kind="hist"
)
plt.show()

```

```python
# KDE

sns.pairplot(
    iris,
    diag_kind="kde" # Diagonal me smooth distribution curve.
)
plt.show()
```

**6. corner=True**

```python
# Upper Triangle remove kar deta hai.
# Duplicate graphs nahi dikhata.

sns.pairplot(
    iris,
    corner=True
)

plt.show()
```

**7. vars Parameter**

```python
# Sirf selected columns compare karna.

sns.pairplot(
    iris,
    vars=[
        "sepal_length",
        "petal_length",
        "petal_width"
    ]
)

plt.show()
```

**8. plot_kws**

```python
# Scatter Plot customize karta hai.
sns.pairplot(
    iris,
    plot_kws={
        "s":80  # Point Size
    }
)

plt.show()
```

**9. diag_kws**

```python
sns.pairplot(
    iris,
    diag_kind="hist",
    diag_kws={
        "bins":15  # histogram -> 15 bins 
        # "bins":5   # histogram -> 5 bins 
    }
)
plt.show()
```

**10. Combined Professional Example**

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

sns.set_theme(style="whitegrid")

sns.pairplot(
    iris,
    hue="species",  # Species Colors.
    diag_kind="kde", # Smooth Distribution.
    corner=True, # Duplicate graphs remove.
    plot_kws={
        "s":60  # Scatter Point Size.
    }
)

plt.show()
```
