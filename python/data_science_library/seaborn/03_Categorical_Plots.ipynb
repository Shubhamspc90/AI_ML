```python
# Notes
# barplot() → Mean compare karta hai.
# countplot() → Frequency compare karta hai.
# boxplot() → Data Summary + Outliers.
# violinplot() → Distribution + Box Plot.
# stripplot() → Individual Points.
# swarmplot() → Individual Points without overlap.
# hue → Category compare.
# palette → Color Theme.
```

**1. What are Categorical Plots?**

```python
# Categorical Plot categories ko compare karta hai.

# Example
# Department      Salary

# HR              40000
# IT              70000
# Sales           55000
# Marketing       50000

# Ya

# Gender
# Male
# Female

# Yaha X-axis categories hoti hain.
```

**2. Dataset Load**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

print(tips.head())
```

**3. barplot()**

```python
# Bar Plot
# ↓
# Average value show karta hai.
# Default me Mean calculate karta hai.

sns.barplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.show()
```

**4. countplot()**

```python
# Count Plot
# ↓
# Har category me kitni values hain.

sns.countplot(
    data=tips,
    x="day"
)
plt.show()
```

**5. boxplot()**

```python
# Box Plot
# ↓
# Data ka summary show karta hai.

# Minimum
# Q1
# Median
# Q3
# Maximum
# Outliers

sns.boxplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.show()
```

**6. violinplot()**

```python
# Violin Plot
# ↓
# Box Plot + Distribution

sns.violinplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.show()
```

**7. stripplot()**

```python
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.show()
```

**8. swarmplot()**

```python
# Strip Plot jaisa hi hai
# Lekin
# Points overlap nahi karte

sns.swarmplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.show()
```

**9. Combined Example**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8,5))

sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    palette="Set2"
)

plt.title("Average Bill by Day and Gender")

plt.show()
```
