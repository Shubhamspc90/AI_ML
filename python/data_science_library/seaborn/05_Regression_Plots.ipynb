```python
# Notes
# regplot() → Scatter + Regression Line
# lmplot() → Regression + Categories
# ci=None → Confidence Interval remove
# scatter=False → Sirf Line
# order=2 → Polynomial Regression
# robust=True → Outlier Resistant Regression
# hue → Multiple Regression Lines
```

**1. What is Regression?**

```python
# Regression ek statistical technique hai jo batati hai ki
# Ek variable dusre variable ko kitna affect karta hai.
# Example

# Study Hours      Marks
# 2                 40
# 3                 48
# 4                 58
# 5                 68
# 6                 80
# 7                 90
# Graph dekhkar pata chal raha hai
# Study Hours ↑
# ↓
# Marks ↑
# Regression Line isi trend ko represent karti hai.
```

**2. Why Regression Plot?**

```python
# Regression Plot use hota hai

# Machine Learning
# Trend Analysis
# Prediction
# Statistics
# Forecasting
```

```python
# Dataset Load
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

print(tips.head())
```

**3. regplot()**

```python
# Regression Plot
# ↓
# Scatter Plot + Regression Line

sns.regplot(
    data = tips,
    x = "total_bill",
    y = "tip"
       
)
plt.show()
```

**4. Confidence Interval (CI)**

```python
# Regression Line ke aas-paas halka shaded area hota hai.
# Ye batata hai
# Regression estimate kitna reliable hai.
# Default
# 95%
# Confidence Interval.

# CI Remove Karna
sns.regplot(
    data=tips,
    x="total_bill",
    y="tip",
    ci=None
)

plt.show()
```

**5. scatter=False**

```python
# Kabhi sirf Regression Line chahiye.

sns.regplot(
    data=tips,
    x="total_bill",
    y="tip",
    scatter=False
)

plt.show()
```

**6. order (Polynomial Regression)**

```python
# Kabhi data straight line follow nahi karta.
# Curve hota hai.
# Us case me
# Polynomial Regression.

sns.regplot(
    data=tips,
    x="total_bill",
    y="tip",
    order=2
)

plt.show()
```

**7. robust=True**

```python
# Outliers hone par normal regression line thodi distort ho sakti hai.
# robust=True outliers ka effect kam karta hai.

sns.regplot(
    data=tips,
    x="total_bill",
    y="tip",
    robust=True
)

plt.show()
```

**8. lmplot()**

```python
# lmplot()
# ↓
# Regression Plot + Category Support
# Ye FacetGrid use karta hai.

sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip"
)
plt.show()
```

```python
# hue Parameter

sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex"
)
plt.show()
```

**9. Combined Professional Example**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.set_theme(style="whitegrid")

sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    height=6,
    aspect=1.4
)

plt.show()
```
