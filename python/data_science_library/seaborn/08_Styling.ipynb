Visualization sirf graph banana nahi hota. Professional graph banana equally important hai.

Isi liye Seaborn itna popular hai.

```python
# set_theme() → Overall Theme.
# set_style() → Background Style.
# set_palette() → Color Scheme.
# figure(figsize=()) → Graph Size.
# title() → Graph Title.
# xlabel()/ylabel() → Axis Labels.
# grid() → Grid Lines.
# despine() → Remove Extra Borders.
# set_context() → Scale for Paper, Notebook, Talk, Poster.
```

**Dataset loading**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
```

**2. set_theme()**

```python
# Entire graph ka overall look change karta hai.
sns.set_theme()
sns.scatterplot(
    data = tips,
    x = "total_bill",
    y = "tip"
)
plt.show()
```

```python
# Dark Grid
sns.set_theme(style="darkgrid")
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.show()
```

```python
# white Grid
sns.set_theme(style="whitegrid")
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.show()
```

```python
# White
sns.set_theme(style="white")
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.show()
```

```python
# Dark
sns.set_theme(style="dark")
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.show()
```

```python
# Ticks
sns.set_theme(style="ticks")
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.show()
```

```python
sns.set_style("whitegrid")

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.show()
```

```python
# Graph ke colors change karta hai.

sns.set_palette("bright")

sns.barplot(
    data=tips,
    x="day",
    y="total_bill"
)

plt.show()
```

**8. Grid**

```python
# Grid graph ko readable banata hai.

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.grid(True)
# plt.grid(False)
plt.show()
```

**9. despine()**

```python
# Extra borders remove karta hai.

sns.barplot(
    data=tips,
    x="day",
    y="total_bill"
)

sns.despine()

plt.show()
```

**11. Combined Styling Example**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.set_theme(style="whitegrid") # Professional Background.
sns.set_palette("deep")  # Professional Colors.
sns.set_context("talk")  # Presentation Size.

plt.figure(figsize=(10,6)) # Large Figure.

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex", # Different Colors.
    style="smoker", # Different Marker Shapes.
    s=120 # Large Markers.
)

plt.title(
    "Restaurant Bill vs Tip",
    fontsize=18,
    fontweight="bold"
)

plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

plt.grid(True)

sns.despine() # Professional Borders.

plt.show()
```

**12. Professional Dashboard Example**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.set_theme(style="whitegrid")
sns.set_context("talk")
sns.set_palette("colorblind")

plt.figure(figsize=(12,6))

sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)

plt.title(
    "Average Total Bill by Day and Gender",
    fontsize=18,
    fontweight="bold"
)

plt.xlabel("Day")
plt.ylabel("Average Bill")

plt.grid(axis="y")

sns.despine()

plt.show()
```
