```python
# scatterplot() → Relationship between two numerical variables.
# lineplot() → Trend show karta hai.
# relplot() → High-level relational plot.
# hue → Colors.
# style → Marker Shapes.
# size → Marker Size.
# col → Multiple Panels.
```

**1. What are Relational Plots?**

```python
# Relational Plot do variables ke beech relationship ko visualize karta hai.

# Example
# Study Hours      Marks
# 2                  40
# 3                  50
# 4                  62
# 5                  75
# 6                  85
# Graph dekhkar turant pata chal jayega
# Study Hours badhne par Marks bhi badh rahe hain.

# Real World Example
# Ek HR Manager dekhna chahta hai
# Experience
# ↓
# Salary

# Ya

# Advertisement Cost
# ↓
# Sales

# Ya

# Temperature
# ↓
# Ice Cream Sales

# Sabme Relational Plot use hota hai.
```

**2. Dataset Load**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

print(tips.head())
```

**3. scatterplot()**

```python
# Scatter Plot
# ↓
# Har observation ko ek point ke form me show karta hai.

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.show()
```

**4. lineplot()**

```python
# Line Plot
# ↓
# Trend dikhata hai.
# Continuous data ke liye use hota hai

sns.lineplot(
    data=tips,
    x="size",
    y="total_bill"
)

plt.show()
```

**5. hue Parameter**

```python
# Different Categories
# ↓
# Different Colors.

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex"
)

plt.show()
```

**6. style Parameter**

```python
# Categories ko Different Marker Shapes milte hain.
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    style="smoker"
)

plt.show()
```

**7. size Parameter**

```python
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    size="size"
)

plt.show()
```

**8. relplot()**

```python
# Relational Plot ka high-level function.
# Ye automatically
# Scatter
# ya
# Line Plot
# create kar sakta hai.

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip"
)

plt.show()

```

**9. Facet Columns**

```python
# Ek hi graph ko multiple panels me divide karta hai.

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    col="sex"
)
plt.show()
```

**10. Combined Professional Example**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.set_theme(style="whitegrid")

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker",
    size="size",
    col="time"
)

plt.show()
```
