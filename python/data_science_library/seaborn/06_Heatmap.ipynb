```python
# Notes
# heatmap() → Color-based visualization.
# corr() → Correlation Matrix.
# annot=True → Values show karta hai.
# cmap → Color Theme.
# linewidths → Cell Borders.
# linecolor → Border Color.
# fmt → Decimal Formatting.
# vmin/vmax → Color Scale.
# isnull() → Missing Value Detection.
```

**1. What is Heatmap?**

```python
# Heatmap ek visualization hai jo values ko different colors se show karta hai.
# High Value → Dark Color
# Low Value → Light Color
# Isse data ko quickly analyze karna easy ho jata hai

# Real World Example
# Ek teacher ke paas 100 students ke marks hain.
# Instead of
# Math = 95
# Science = 80
# English = 60

# Heatmap me
# 🟩 95
# 🟨 80
# 🟥 60
# Ek glance me samajh aa jayega kis subject me performance achhi hai.
```

**2. Why Heatmap?**

```python
# Heatmap use hota hai

# Correlation Analysis
# Feature Selection
# Missing Value Detection
# Data Analysis
# Business Dashboard
# Machine Learning
```

```python
# Dataset Load
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

print(tips.head())
```

**3. Correlation Matrix**

```python
# Correlation batata hai
# Do Numerical Columns kitne related hain.
# Range hoti hai

# +1   → Strong Positive Correlation
#  0   → No Correlation
# -1   → Strong Negative Correlation

corr = tips.corr(numeric_only=True)

print(corr)
```

**4. Basic Heatmap**

```python
corr = tips.corr(numeric_only=True)

sns.heatmap(corr) # Correlation Matrix ko Heatmap me convert karta hai.

plt.show()
```

**5. annot=True**

```python
# Cells ke andar values bhi show karega.
corr = tips.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True  # Har cell ke andar value print karega.
)

plt.show()
```

**6. cmap**

```python
# Color Theme change karta hai.
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)
plt.show()

# Popular Color Maps
# cmap="Blues"
# cmap="Greens"
# cmap="Reds"
# cmap="viridis"
# cmap="magma"
# cmap="YlGnBu"
```

**7. linewidths**

```python
# Cells ke beech line ki thickness.
sns.heatmap(
    corr,
    annot=True,
    linewidths=2
)

plt.show()
```

**8. linecolor**

```python
# Cell Border Color.

sns.heatmap(
    corr,
    annot=True,
    linewidths=1,
    linecolor="black"
)

plt.show()
```

**9. fmt**

```python
# Numbers kis format me show honge.
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f"  # 2 Decimal Places.
)

plt.show()
```

**10. vmin & vmax**

```python
# Color Range manually set karna.
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    vmin=-1,
    vmax=1
)

plt.show()
```

**11. Missing Value Heatmap**

```python
# Missing Values identify karne ke liye bhi Heatmap use hota hai.
import pandas as pd

df = pd.read_csv("../datasets/students.csv")

sns.heatmap(
    df.isnull(),
    cbar=False
)

plt.show()
```

**12. Combined Professional Example**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

corr = tips.corr(numeric_only=True)

plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    linewidths=1,
    linecolor="white",
    fmt=".2f",
    vmin=-1,
    vmax=1
)

plt.title("Correlation Heatmap")

plt.show()
```
