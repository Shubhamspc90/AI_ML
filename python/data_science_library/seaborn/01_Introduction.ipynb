**4. Import Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**5. Built-in Datasets**

```python
# Seaborn ke paas kuch ready-made datasets hote hain.
sns.get_dataset_names()
```

```python
# Iris Dataset Load Karna hai
iris = sns.load_dataset("iris")  # Internet se built-in dataset load karta hai.

print(iris.head())
```

**6. First Seaborn Plot**

```python
iris=sns.load_dataset("iris")
data=sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length"
)

plt.show()
```

**7. Seaborn vs Matplotlib**

```python

# | Matplotlib          | Seaborn                   |
# | ------------------- | ------------------------- |
# | Basic Visualization | Statistical Visualization |
# | More Code           | Less Code                 |
# | Manual Styling      | Automatic Styling         |
# | Basic Themes        | Professional Themes       |
# | Low-Level Library   | High-Level Library        |

# Matplotlib
plt.scatter(
    iris["sepal_length"],
    iris["petal_length"]
)

plt.show()
```

```python
# Seaborn
# Seaborn ka code zyada clean aur readable hai.
sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length"
)

plt.show()
```

**8. Themes**

```python
# Seaborn me built-in themes hoti hain:
# sns.set_theme(style="darkgrid")
# sns.set_theme(style="dark")
# sns.set_theme(style="white")
# sns.set_theme(style="ticks")

sns.set_theme(style="whitegrid")
sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length"
)

plt.show()
```

**9. Color Palette**

```python
# sns.set_palette("deep") # Default colors ke alawa
# sns.set_palette("pastel")
# sns.set_palette("bright")
# sns.set_palette("dark")
# sns.set_palette("colorblind")

sns.set_theme(style="whitegrid")
# sns.set_palette("bright")
sns.set_palette("dark")

sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length"
)

plt.show()
```

**10. Combined Example**

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

sns.set_theme(style="whitegrid")
sns.set_palette("deep")

sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species",
    s=100
)

plt.title("Iris Dataset Visualization")

plt.show()
```
