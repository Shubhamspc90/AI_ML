```python
# Notes
# histplot() → Histogram
# bins → Number of Groups
# kde=True → Smooth Curve
# kdeplot() → Density Curve
# fill=True → Fill Color
# hue → Category Comparison
# rugplot() → Individual Data Points
# ecdfplot() → Cumulative Distribution
```

**1. What is Distribution?**

```python
# Distribution batata hai

# Data kis range me hai
# Sabse jyada values kaha hain
# Data evenly spread hai ya nahi
# Outliers hain ya nahi
# Data Normal Distribution follow karta hai ya nahi
# Data Science me model banane se pehle Distribution dekhna bahut important hota hai.

# Real World Example
# Ek teacher ke paas 100 students ke marks hain.
# Distribution dekhkar pata chal sakta hai
# 0-20      2 Students
# 20-40     10 Students
# 40-60     35 Students
# 60-80     40 Students
# 80-100    13 Students
```

**2. Why Distribution Plots?**

```python
# Distribution Plots ka use hota hai:

# Data Analysis
# Machine Learning
# Statistics
# Data Cleaning
# Outlier Detection
# Feature Engineering


# Dataset Load
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")  # Ye Seaborn ka built-in restaurant dataset hai.
print(tips.head())
```

**3. histplot()**

```python
# Histogram ki tarah kaam karta hai.
# Frequency show karta hai.

sns.histplot(
    data=tips,
    x="total_bill"
)
plt.show()
```

**4. bins Parameter**

```python
# Histogram kitne groups banayega.
sns.histplot(
    data=tips,
    x="total_bill",
    bins=5
)
plt.show()
```

```python
sns.histplot(
    data=tips,
    x="total_bill",
    bins=20
)
plt.show()
```

**5. KDE ( Kernel Density Estimation )**

```python
# Ye Histogram ke upar ek smooth curve banata hai.

sns.histplot(
    data = tips,
    x="total_bill",
    kde=True
)
plt.show()
```

**6. kdeplot()**

```python
# Sirf Curve dikhana ho
sns.kdeplot(
    data=tips,
    x="total_bill"
)
plt.show()
```

**7. fill=True**

```python
# Curve ke niche color fill karta hai.
sns.kdeplot(
    data=tips,
    x="total_bill",
    fill=True
)
plt.show()
```

**8. Multiple Distribution**

```python
# Do categories compare karna.

sns.histplot(
    data=tips,
    x="total_bill",
    hue="sex"
)

plt.show()
```

**9. rugplot()**

```python
sns.rugplot(
    data=tips,
    x="total_bill"
)

plt.show()
```

**10. ecdfplot()**

```python
# ECDF=Empirical Cumulative Distribution Function
# Batata hai
# Kitna percentage data kisi value se kam hai.
sns.ecdfplot(
    data=tips,
    x="total_bill"
)

plt.show()

```

**11. Combined Example**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.set_theme(style="whitegrid")

plt.figure(figsize=(9,5))

sns.histplot(
    data=tips,
    x="total_bill",
    bins=15,
    kde=True,
    color="royalblue"
)

plt.title("Distribution of Total Bill")

plt.xlabel("Total Bill")

plt.ylabel("Frequency")

plt.show()
```
