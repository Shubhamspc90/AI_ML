```python
# # Principal Component Analysis (PCA)

# ## What is PCA?

# Principal Component Analysis (PCA) is an Unsupervised Learning technique used for Dimensionality Reduction.

# It reduces the number of features while preserving as much important information as possible.

# Instead of using all original features, PCA creates new features called Principal Components.

# ## Why PCA?

# - Reduces Dataset Size
# - Faster Model Training
# - Removes Redundant Features
# - Reduces Noise
# - Helps in Data Visualization

# ## Real World Applications

# - Face Recognition
# - Image Compression
# - Recommendation Systems
# - Medical Data Analysis
# - Finance
```

Code Cell 1  Load Dataset

```python
# We'll use the Wine dataset because it contains many numerical features.

from sklearn.datasets import load_wine
wine  = load_wine()
X=wine.data
print("Datasets: ",X.shape)

```

```python
# Dataset Shape : (178, 13)
# Meaning
# 178 samples
# 13 features
```

```python
# # Why Feature Scaling?

# PCA is based on variance.

# Features with larger values dominate the principal components.

# Therefore, feature scaling is required before applying PCA.
```

Code Cell 2

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
# Explanation

# Now every feature has

# Mean = 0
# Standard Deviation = 1

# This ensures that all features contribute equally.
```

Code Cell 3
 Apply PCA

```python
from sklearn.decomposition import PCA

pca = PCA(

    n_components=2

)

X_pca = pca.fit_transform(

    X_scaled

)
```

```python
# Explanation

# n_components=2
# means

# Reduce
# 13 Features
# ↓
# 2 Principal Components
```

Code Cell 4
 Check Shape

```python
print("Original Shape :", X.shape)

print("Reduced Shape :", X_pca.shape)
```

Code Cell 5 (imp.)
 Explained Variance Ratio

```python
print(pca.explained_variance_ratio_)
```

```python
# Explanation
# Suppose output is
# [0.36 0.19]

# Meaning

# First Component
# ↓
# 36% Information

# Second Component
# ↓
# 19% Information

# Together
# ↓
# 55% Information
```

Code Cell 6 (imp.) Total Explained Variance

```python
print(pca.explained_variance_ratio_.sum())
```

```python
# Explanation

# This tells us how much information is retained after
# dimensionality reduction.
```

Code Cell 7
 Visualize PCA

```python
import matplotlib.pyplot as plt

plt.scatter(

    X_pca[:,0],

    X_pca[:,1]

)

plt.xlabel("Principal Component 1")

plt.ylabel("Principal Component 2")

plt.title("PCA Visualization")

plt.show()
```

```python
# Explanation

# Each point represents one wine sample.

# The axes are no longer original features.

# They are Principal Components.
```

Code Cell 8
 Create DataFrame

```python
import pandas as pd

df = pd.DataFrame(

    X_pca,

    columns=[

        "PC1",

        "PC2"

    ]

)

df.head()
```

```python
# # How PCA Works?

# Original Features

# ↓

# Calculate Variance

# ↓

# Find Principal Components

# ↓

# Sort by Importance

# ↓

# Keep Top Components

# ↓

# Reduced Dataset
```

```python
# Real World Example

# Suppose an employee dataset has

# Age

# Salary

# Experience

# Education

# City

# Department

# Gender

# Projects

# Performance

# Attendance

# Bonus

# Working Hours

# Training Hours

# Instead of using all 13 features,

# PCA may reduce them to

# Principal Component 1

# Principal Component 2

# while keeping most of the useful information.
```

Advantages

```python
# Faster Training
# Less Memory
# Removes Noise
# Better Visualization
# Handles High-Dimensional Data
```

Limitations

```python
# Components are difficult to interpret.
# Some information is always lost.
# Requires feature scaling.
```
