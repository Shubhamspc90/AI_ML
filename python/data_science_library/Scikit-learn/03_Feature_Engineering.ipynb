```python
# # Feature Engineering

# ## What is Feature Engineering?

# Feature Engineering is the process of transforming raw data into a format that Machine Learning algorithms can understand and learn from.

# Good features lead to better models.

# ## Why is Feature Engineering Important?

# - Improves model performance
# - Converts categorical data into numerical form
# - Removes unnecessary information
# - Makes data suitable for Machine Learning

# ## Topics Covered

# - One-Hot Encoding
# - Ordinal Encoding
# - Feature Selection
# - fit()
# - transform()
# - fit_transform()
```

Code Cell 1

```python
import pandas as pd

df = pd.DataFrame({

    "City":["Delhi","Mumbai","Delhi","Pune","Mumbai"],

    "Education":["UG","PG","UG","PhD","PG"],

    "Salary":[30000,50000,35000,70000,60000]

})

df
```

Code Cell 2
One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(

    sparse_output=False # OneHotEncoder returns a Sparse Matrix.

)

city = encoder.fit_transform(

    df[["City"]]

)

city
```

Code Cell 3
Column Names

```python
city_df = pd.DataFrame(

    city,

    columns=encoder.get_feature_names_out()

)

city_df
```

Code Cell 4
Merge with Original Dataset

```python
df = pd.concat(

    [

        df.drop("City", axis=1),

        city_df

    ],

    axis=1

)

df
```

Code Cell 5
Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder(

    categories=[

        [

            "UG",

            "PG",

            "PhD"

        ]

    ]

)

df["Education"] = encoder.fit_transform(

    df[["Education"]]

)

df
```

Code Cell 6
Understanding fit()

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(

    df[["Salary"]]

)
```

Code Cell 7
Understanding transform()

```python
scaled = scaler.transform(

    df[["Salary"]]

)

scaled
```

Code Cell 8
Understanding fit_transform()

```python
scaler = StandardScaler()

salary = scaler.fit_transform(

    df[["Salary"]]

)

salary
```
