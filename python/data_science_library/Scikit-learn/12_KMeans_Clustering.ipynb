```python
# # K-Means Clustering

# ## What is K-Means?

# K-Means is an Unsupervised Machine Learning algorithm.

# Unlike Supervised Learning, K-Means does not require target labels.

# It automatically groups similar data points into different clusters.

# ## Real World Applications

# - Customer Segmentation
# - Market Analysis
# - Product Recommendation
# - Image Compression
# - Document Clustering

# ## What You Will Learn

# - Clustering
# - Centroids
# - K Value
# - Elbow Method
# - Cluster Prediction
# - Cluster Visualization
```

Code Cell 1 Load a sample dataset.

```python
# We'll use the Wine dataset and ignore the target values
# because K-Means is an Unsupervised Learning algorithm

from sklearn.datasets import load_wine
wine = load_wine()
X=wine.data
print("dataset shape : ",X.shape)
```

```python
# # Why Feature Scaling?

# K-Means uses distance to find similar data points.

# Features with larger values can dominate the distance calculation.

# Therefore, feature scaling is very important.
```

Code Cell 2

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Explanation
# Every feature now has
# Mean = 0
# Standard Deviation = 1
# Now every feature contributes equally.
```

Code Cell 3
 Create K-Means Model

```python
from sklearn.cluster import KMeans
model = KMeans(
    n_clusters=3, # Create 3 Groups
    random_state=42
)
```

Code Cell 4 (imp. )Train the Model

```python
model.fit(
    X_scaled
)

# Explanation
# Unlike previous algorithms,
# there is no y (target).
# The algorithm simply looks for similar data points.
```

Code Cell 5
 Cluster Labels

```python
clusters = model.labels_

clusters[:20]
```

```python
# Explanation
# Example output
# 0
# 2
# 1
# 2
# 0
# Each number represents the cluster assigned to that data point.
```

Code Cell 6
Predict Cluster for New Data

```python
prediction = model.predict(

    X_scaled[:5]

)

print(prediction)
```

Code Cell 7
 Centroids

```python
print(model.cluster_centers_)

# Explanation

# Centroids are the center points of each cluster.
# Every cluster has one centroid.
```

Code Cell 8  Visualize Clusters

```python
import matplotlib.pyplot as plt

plt.scatter(

    X_scaled[:,0],

    X_scaled[:,1],

    c=clusters

)

plt.xlabel("Feature 1")

plt.ylabel("Feature 2")

plt.title("K-Means Clusters")

plt.show()

# Different colors represent different clusters.
```

```python
# # How K-Means Works?

# Choose K

# ↓

# Select Random Centroids

# ↓

# Assign Data Points

# ↓

# Update Centroids

# ↓

# Repeat Until Centroids Stop Moving

# ↓

# Final Clusters
```

Code Cell 9 
Elbow Method

```python
inertia = []
for k in range(1,11):
    model = KMeans(
        n_clusters=k,
        random_state=42
    )
    
    model.fit(X_scaled)
    inertia.append(
        model.inertia_
    )
    
# Explanation

# inertia_ measures how compact the clusters are.
# Lower inertia means tighter clusters.
```

Code Cell 10
Plot Elbow Graph

```python
import matplotlib.pyplot as plt

plt.plot(

    range(1,11),

    inertia,

    marker="o"

)

plt.xlabel("Number of Clusters")

plt.ylabel("Inertia")

plt.title("Elbow Method")

plt.show()

# Explanation

# Look for the point where the curve bends like an elbow.
# That value is usually the best choice for K.
```

```python
# Real World Example

# Imagine a shopping mall with 10,000 customers.
# The mall owner doesn't know who belongs to which customer type.
# K-Means automatically creates groups like:

# Cluster 1
# ↓
# High Income
# High Spending

# Cluster 2
# ↓
# Medium Income
# Medium Spending

# Cluster 3
# ↓
# Low Income
# Low Spending

# Now the company can create different marketing strategies for each customer group.
```
