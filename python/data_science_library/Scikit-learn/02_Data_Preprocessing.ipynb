```python
# # Data Preprocessing

# ## What is Data Preprocessing?

# Data Preprocessing is the process of cleaning and preparing raw data before training a Machine Learning model.

# Real-world data is rarely clean. It often contains:

# - Missing Values
# - Duplicate Data
# - Categorical Values
# - Different Scales
# - Incorrect Data Types

# If we train a model on unclean data, the model's performance will be poor.

# ## Why is Data Preprocessing Important?

# - Improves Model Accuracy
# - Makes Data Consistent
# - Reduces Errors
# - Helps Algorithms Learn Better

# ## Steps Covered

# - Missing Values
# - Encoding
# - Feature Scaling
```

Code Cell 1

```python
# ============================================================
# Import Libraries
# ============================================================

import pandas as pd
import numpy as np
```

Code Cell 2
Create Sample Dataset

```python
df = pd.DataFrame({

    "Age":[22,25,np.nan,30,28,np.nan],

    "Salary":[25000,30000,28000,np.nan,45000,50000],

    "Gender":["Male","Female","Female","Male","Female","Male"]

})

df
```

Code Cell 3
Check Missing Values

```python
df.isnull()

# True → Missing Value

# False → Available Value
```

Code Cell 4
Count Missing Values

```python
df.isnull().sum()
```

Code Cell 5
Fill Missing Values (Mean)

```python
df["Age"]=df["Age"].fillna(df["Age"].mean())
df["Salary"]=df["Salary"].fillna(df["Salary"].mean())
df
```

Code Cell 6
Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df["Gender"] = encoder.fit_transform(df["Gender"])
df
```

Code Cell 7
Standard Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[["Age","Salary"]] = scaler.fit_transform(

    df[["Age","Salary"]]

)

df
```

Code Cell 8
Min-Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler

df = pd.DataFrame({

    "Age":[22,25,28,30,35]

})

scaler = MinMaxScaler()

df["Age"] = scaler.fit_transform(df[["Age"]])

df
```

📝 Machine Learning Flow

```python
# Raw Data
# ↓
# Missing Values
# ↓
# Encoding
# ↓
# Scaling
# ↓
# Train-Test Split
# ↓
# Machine Learning Model
```
