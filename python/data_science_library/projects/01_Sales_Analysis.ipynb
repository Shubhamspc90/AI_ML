```python
# # Step 10 : Data Cleaning

# Now we will clean missing values.
```

**Code Cell 14**

```python
# ============================================================
# Fill Customer Rating
# ============================================================

df["Customer_Rating"] = df["Customer_Rating"].fillna(
    df["Customer_Rating"].median()
)

print(df["Customer_Rating"].isnull().sum())
```

**Code Cell 15**

```python
# ============================================================
# Fill Discount
# ============================================================

df["Discount"] = df["Discount"].fillna(
    df["Discount"].median()
)

print(df["Discount"].isnull().sum())
```

**Code Cell 16**

```python
# ============================================================
# Fill Email
# ============================================================

df["Email"] = df["Email"].fillna("Not Available")

print(df["Email"].isnull().sum())
```

**Code Cell 17**

```python
# ============================================================
# Fill Phone
# ============================================================

df["Phone"] = df["Phone"].fillna("Not Available")

print(df["Phone"].isnull().sum())
```

**Code Cell 18**

```python
# ============================================================
# Verify Missing Values
# ============================================================

print("="*60)
print("Remaining Missing Values")
print("="*60)

print(df.isnull().sum())
```

**Code Cell 19**

```python
# ============================================================
# Invalid Age
# ============================================================

invalid_age = df[df["Age"] < 0]

print("="*60)
print("Invalid Age Records")
print("="*60)

print(invalid_age.shape)

invalid_age.head()
```

**Code Cell 20**  ( Fix Invalid Ages )

```python
# ============================================================
# Replace Invalid Age
# ============================================================

median_age = df[df["Age"] > 0]["Age"].median()

df.loc[df["Age"] < 0, "Age"] = median_age

print(df[df["Age"] < 0].shape)
```

```python
# # Step 15 : Time Series Analysis

# Time Series Analysis helps us understand how business changes over time.

# Objectives:

# - Monthly Sales Trend
# - Monthly Profit Trend
# - Yearly Growth
# - Quarterly Analysis
# - Weekend vs Weekday Sales
# - Top Sales Months
```

**Code Cell 54**

```python
# ============================================================
# Convert Order Date
# ============================================================

df["Order_Date"] = pd.to_datetime(df["Order_Date"])

print(df["Order_Date"].dtype)

# datetime64[ns] ka matlab
# datetime64 → Date aur Time store karta hai.
# ns → Nanosecond precision.
# Iska matlab Pandas ka Date-Time data type hai.
```

**Code Cell 55**

```python
# ============================================================
# Create Date Features
# ============================================================

df["Year"] = df["Order_Date"].dt.year

df["Month"] = df["Order_Date"].dt.month

df["Month_Name"] = df["Order_Date"].dt.month_name()

df["Quarter"] = df["Order_Date"].dt.quarter

df["Weekday"] = df["Order_Date"].dt.day_name()

print(df[[
    "Order_Date",
    "Year",
    "Month",
    "Month_Name",
    "Quarter",
    "Weekday"
]].head())
```

**Code Cell 56**

```python
# ============================================================
# Monthly Sales Trend
# ============================================================

monthly_sales = (
    df
    .groupby(["Year","Month"])["Sales"]
    .sum()
    .reset_index()
)

print(monthly_sales.head())
plt.figure(figsize=(14,6))

plt.plot(
    monthly_sales.index,
    monthly_sales["Sales"],
    marker="o"
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month Index")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# Business Insight
# Business growth over time
# Seasonal trends
# Peak months
```

**Code Cell 57**

```python
# ============================================================
# Monthly Profit Trend
# ============================================================

monthly_profit = (
    df
    .groupby(["Year","Month"])["Profit"]
    .sum()
    .reset_index()
)

plt.figure(figsize=(14,6))
plt.plot(
    monthly_profit.index,
    monthly_profit["Profit"],
    marker="o"
)

plt.title("Monthly Profit Trend")
plt.xlabel("Month Index")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```

**Code Cell 58**

```python
# ============================================================
# Year Wise Sales
# ============================================================

year_sales = (
    df
    .groupby("Year")["Sales"]
    .sum()
)
print(year_sales)

plt.figure(figsize=(8,5))
year_sales.plot(kind="bar")
plt.title("Year Wise Sales")
plt.ylabel("Sales")

plt.show()
```

```python
# Business Insight
# Compare yearly growth
# Identify business expansion
```

**Code Cell 60**

```python
# ============================================================
# Month Wise Sales
# ============================================================

month_order = [
    "January","February","March","April",
    "May","June","July","August",
    "September","October","November","December"
]

month_sales = (
    df
    .groupby("Month_Name")["Sales"]
    .sum()
    .reindex(month_order)
)

plt.figure(figsize=(12,5))
month_sales.plot(kind="bar")
plt.title("Month Wise Sales")
plt.ylabel("Sales")
plt.xticks(rotation=45)

plt.show()
```

**Code Cell 61**

```python
# ============================================================
# Weekend vs Weekday
# ============================================================

weekend = ["Saturday", "Sunday" ]

df["Day_Type"] = np.where(
    df["Weekday"].isin(weekend),
    "Weekend",
    "Weekday"
)

day_sales = (
    df
    .groupby("Day_Type")["Sales"]
    .sum()
)
print(day_sales)

plt.figure(figsize=(6,5))
day_sales.plot(kind="bar")
plt.title("Weekend vs Weekday Sales")

plt.show()
```
