```python
# Notes
# plt.subplot() → Old method
# plt.subplots() → Recommended method
# fig → Figure object
# ax → Axes object
# sharex=True → Same X-axis
# sharey=True → Same Y-axis
# tight_layout() → Overlap remove karta hai
```

**1. What are Subplots?**

```python
# Subplots ka matlab hai:

# Ek hi Figure ke andar multiple graphs banana.
```

**2. Why use Subplots?**

```python
# Subplots ka use hota hai:

# Dashboard banane me
# Data Comparison
# Reports
# Data Analysis
# Machine Learning Visualization
```

**3. plt.subplot()**

```python
# Syntax
# plt.subplot(rows, columns, position)
# Meaning:
# rows → Kitni rows hongi
# columns → Kitne columns honge
# position → Graph ki position
```

**4. First Subplot**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]

y = [10,20,15,25]

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)

plt.plot(x,y)

plt.title("Line Plot")

plt.subplot(1,2,2)

plt.bar(x,y)

plt.title("Bar Chart")

plt.show()
```

**5. 2 Rows 2 Columns**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]

y = [10,20,15,25]

plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Line")

plt.subplot(2,2,2)
plt.bar(x,y)
plt.title("Bar")

plt.subplot(2,2,3)
plt.scatter(x,y)
plt.title("Scatter")

plt.subplot(2,2,4)
plt.hist(y)
plt.title("Histogram")

plt.show()
```

**6. plt.subplots()**

```python
# Ye modern aur recommended method hai.
import matplotlib.pyplot as plt

x = [1,2,3,4]

y = [10,20,15,25]

fig, ax = plt.subplots()

ax.plot(x,y)

plt.show()
```

**7. Multiple Axes**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]

y = [10,20,15,25]

fig, ax = plt.subplots(2,2)

ax[0,0].plot(x,y)
ax[0,0].set_title("Line")

ax[0,1].bar(x,y)
ax[0,1].set_title("Bar")

ax[1,0].scatter(x,y)
ax[1,0].set_title("Scatter")

ax[1,1].hist(y)
ax[1,1].set_title("Histogram")

plt.show()
```

**8. Shared Axes**

```python
# Jab multiple graphs ka same X-axis ya Y-axis ho.
fig, ax = plt.subplots(
    2,
    1,
    sharex=True
)

fig, ax = plt.subplots(
    1,
    2,
    sharey=True
)

# Benefits
# Same scale
# Easy comparison
# Professional reports
```

**9. Tight Layout**

```python
fig, ax = plt.subplots(2,2)

plt.tight_layout()

plt.show()
```

**10. Combined Example**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]

y = [10,20,15,30,25]

fig, ax = plt.subplots(2,2, figsize=(10,8))

ax[0,0].plot(x,y)
ax[0,0].set_title("Line Plot")

ax[0,1].bar(x,y)
ax[0,1].set_title("Bar Chart")

ax[1,0].scatter(x,y)
ax[1,0].set_title("Scatter Plot")

ax[1,1].hist(y)
ax[1,1].set_title("Histogram")

plt.tight_layout()

plt.show()
```
