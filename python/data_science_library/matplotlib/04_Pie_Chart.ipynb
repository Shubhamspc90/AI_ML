```python
# Notes
# pie() → Pie Chart
# labels → Category names
# colors → Slice colors
# autopct → Percentage
# startangle → Rotation
# explode → Highlight slice
# shadow → Shadow effect
# radius → Pie size
# Circle() → Donut Chart
```

**1. What is Pie Chart?**

```python
# Pie Chart ka use percentage ya proportion dikhane ke liye hota hai.
# Isme poora circle 100% ko represent karta hai.
# Har slice kisi category ka percentage batati hai.
```

**2. Basic Pie Chart**

```python
import matplotlib.pyplot as plt
subjects= ["Maths","Science","English","Hindi"]
marks = [90,80,70,60]
plt.pie(marks)
plt.show()
```

**3. Basic Pie Chart**

```python
plt.pie(
    marks,
    labels=subjects
)

plt.show()
```

**4. Colors**

```python
# Default colors automatic hote hain.
plt.pie(
    marks,
    labels=subjects,
    colors=["red","blue","green","orange"]
)

plt.show()
```

**5. Percentage**

```python
plt.pie(
    marks,
    labels=subjects,
    autopct="%1.1f%%"
)

plt.show()
```

**6. Start Angle**

```python
plt.pie(
    marks,
    labels=subjects,
    # startangle=90,
    startangle=180
)

plt.show()
```

**7. Explode**

```python
# Kisi ek slice ko highlight karna.

plt.pie(
    marks,
    labels=subjects,
    explode=(0.2,0,0,0)
    # Agar 2 slices bahar nikalni ho
    # explode=(0.1,0.2,0,0)
)

plt.show()
```

**8. Shadow**

```python
# 3D effect.

plt.pie(
    marks,
    labels=subjects,
    shadow=True
)

plt.show()
```

**9. Radius**

```python
# Pie ka size.
plt.pie(
    marks,
    labels=subjects,
    # radius=1.5
    # Small Pie
    radius=0.8
)

plt.show()
```

**10. Combined Example**

```python

subjects = ["Math", "Science", "English", "Hindi"]
marks = [90,80,70,60]

plt.figure(figsize=(7,7))
plt.pie(
    marks,
    labels=subjects,
    colors=["red","blue","green","orange"],
    autopct="%1.1f%%",
    startangle=90,
    explode=(0.1,0,0,0),
    shadow=True
)
plt.title("Student Marsk Distribution")
plt.show()
```

**11. Donut Chart**

```python
# Pie Chart ka modern version.

import matplotlib.pyplot as plt

subjects = ["Math","Science","English","Hindi"]

marks = [90,80,70,60]

plt.pie(
    marks,
    labels=subjects,
    autopct="%1.1f%%"
)

centre_circle = plt.Circle(
    (0,0),
    0.60,
    fc="white"
)

plt.gca().add_artist(centre_circle)

plt.title("Donut Chart")

plt.show()
```
