```python
# Notes
# plot() → Line graph
# color → Line color
# linestyle → Line style
# linewidth → Line thickness
# marker → Data point
# markersize → Marker size
# label → Legend name
# legend() → Label show karta hai
# grid() → Reading easy banata hai
```

**1. Basic Line Plot**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]

y = [10, 20, 15, 25, 30]

plt.plot(x, y)

plt.show()
```

**2. Multiple Line Plot**

```python
import matplotlib.pyplot as plt
days =[ 1,2,3,4,5]
product_A=[10,20,15,30,25]
product_B=[12,18,20,28,35]
plt.plot(days,product_A)
plt.plot(days,product_B)
plt.show()

```

**3️. Line Color**

```python
plt.plot(days, product_A, color="red")

plt.show()

# Common Colors
# blue
# red
# green
# black
# orange
# yellow
# purple
# pink
# brown
# gray
# cyan
```

**4️. Line Style**

```python
# Default:
# ──────

# Dashed
plt.plot(days, product_A, linestyle="--")
plt.show()


```

```python
# Dotted
plt.plot(days,product_A,linestyle=":")
plt.show()

```

```python
#Dash Dot
plt.plot(days,product_A,linestyle="-.",color="blue")
plt.show()
```

```python
#solid
plt.plot(days, product_A, linestyle="-")
```

**5️. Line Width**
Width badhne se line thick ho jati hai.

```python
plt .plot(days,product_A,linewidth=5)
plt.show()
```

**6️. Marker**
Marker har data point ko highlight karta hai.

```python
# Circle
plt.plot(days,product_A,marker="o")
plt.show()
```

```python
# Square
plt.plot(days,product_A,marker="s")
plt.show()
```

```python
# Star
plt.plot(days, product_A, marker="*")
plt .show()
```

```python
# Triangle
plt.plot(days, product_A, marker="^")
plt.show()
```

```python
# Diamond
plt.plot(days, product_A, marker="D")
plt.show()
```

**7️. Marker Size**

```python
plt.plot(days, product_A, marker="o", markersize=12)

plt.show()
```

**8️. Marker Color**

```python
plt.plot(
    days,
    product_A,
    marker="o",
    markersize=12,
    markerfacecolor="red",
    markeredgecolor="black"
)

plt.show()
```

**9️. Label**

```python
# Label legend ke liye use hota hai.
plt.plot(days, product_A, label="Product A")
plt.show()

#label= sirf label assign karta hai, usko display nahi karta.
#Label ko display karne ke liye plt.legend() zaroori hai.
```

**10. Legend**

```python
plt.plot(days,product_A,label="Prodect A")
plt.plot(days,product_B,label="Prodect B")
plt.legend()
plt.show()
```

**11. Grid Customization**

```python

days = [1, 2, 3, 4, 5]
sales = [100, 150, 120, 180, 200]

plt.plot(days, sales,
         marker="o",
         linewidth=2,
         label="Sales")

plt.title("Daily Sales Report")
plt.xlabel("Days")
plt.ylabel("Sales")

plt.grid(
    True,
    color="gray",
    linestyle="--",
    linewidth=1,
    alpha=0.6
)

plt.legend()

plt.show()
```

**12. Combined Example**

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]

sales = [50, 70, 65, 90, 100]

plt.figure(figsize=(8,5))

plt.plot(
    months,
    sales,
    color="blue",
    linestyle="--",
    linewidth=3,
    marker="o",
    markersize=8,
    label="Sales"
)

plt.title("Monthly Sales")

plt.xlabel("Months")

plt.ylabel("Sales")

plt.grid(alpha=0.5)

plt.legend()

plt.show()
```
