**4. Import Matplotlib**

```python
import matplotlib.pyplot as plt
```

**5. First Plot**

```python
x = [1,2,3,4,5]
y = [10,30,40,60,70]
plt.plot(x,y)
plt.show()
```

**6. Figure**

```python
plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.show()
```

**7️. Axes**

```python
# Axes matlab graph ki lines.
#         Y-axis
#           ↑
#           │
#           │
# ──────────┼────────→ X-axis
```

**8. Labels**

```python
plt.plot(x,y)
plt.xlabel("Months")
plt.ylabel("Sales")

plt.show()
```

**9.Title**

```python
plt.plot(x,y)
plt.title("Monthy sales")
plt.show()
```

**10. Grid**

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

**11. Complete Example**

```python
import matplotlib.pyplot as plt

months=["Jan","Feb","Mar","Apr","May"]
sales=[50,70,65,90,100]
plt.figure(figsize=(8,5))
plt.plot(months,sales)
plt.title("Monthly Sales Report")
plt.xlabel("months")
plt.ylabel("Sales")
plt.show()
```

```python
# Notes
# plt.plot() → Line graph
# plt.show() → Graph display
# plt.figure() → Figure size
# xlabel() → X-axis name
# ylabel() → Y-axis name
# title() → Graph title
# grid() → Grid lines
```
