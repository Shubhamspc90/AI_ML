```python
# Notes
# hist() → Histogram
# bins → Number of groups
# color → Bar color
# edgecolor → Border color
# alpha → Transparency
# density=True → Probability Distribution
```

**1. What is Histogram?**

```python
# Theory

# Histogram continuous numerical data ka frequency distribution dikhata hai.
# Ye batata hai ki
# Kitne students ke marks 0-20 ke beech hain
# Kitne students ke marks 20-40 ke beech hain
# Kitne students ke marks 40-60 ke beech hain
# etc.
# Histogram automatically values ko groups (bins) me divide karta hai.

# Real World Example
# Suppose class ke marks:
# 45
# 55
# 67
# 78
# 82
# 91
# 38
# 49
# 60
# 71

# Histogram batayega

# 0-20      0 Students

# 20-40     1 Student

# 40-60     4 Students

# 60-80     3 Students

# 80-100    2 Students
```

**2. Histogram vs Bar Chart**

```python
#   Histogram	           Bar Chart
# Continuous          Data	Categorical Data
# Bars touch          each other	Bars ke beech gap hota hai
# Frequency           show karta hai	Comparison show karta hai
# Example: Marks	  Example: Subjects
```

**3. Basic Histogram**

```python
import matplotlib.pyplot as plt

marks = [45,55,67,78,82,91,38,49,60,71]

plt.hist(marks)

plt.show()
```

**4. bins Parameter**

```python
# Histogram values ko groups me divide karta hai.
# Ye groups bins kehlate hain.

# Default
plt.hist(marks)
plt.show()
```

```python
# Custom bins
plt.hist(marks, bins=5) # Ab graph sirf 5 groups banayega.

plt.show()
```

```python
# Aur zyada bins
plt.hist(marks, bins=20) # More bins = More detailed graph

plt.show()
```

**5. Color**

```python
plt.hist(marks,color="green")
plt.show()
```

**6. Edge Color**

```python
plt.hist(
    marks,
    edgecolor="black"
)
plt.show()
```

**7. Alpha (Transparency)**

```python
# 0 → Invisible
# 1 → Fully Visible

plt.hist(
    marks,
    # alpha=0.5
    # alpha=1 # Dark
    alpha=0.3 # Light
)

plt.show()
```

**8. Density Histogram**

```python
# Normally histogram frequency show karta hai.
# Density=True
# ↓
#Probability distribution show karta hai

plt.hist(
    marks,
    density=True #Machine Learning aur Statistics me bahut use hota hai.
)
plt.show()
```

**9. Combined Example**

```python
import matplotlib.pyplot as plt

marks = [45,55,67,78,82,91,38,49,60,71,
         65,59,74,88,92,56,61,70,83,95]

plt.figure(figsize=(8,5))

plt.hist(
    marks,
    bins=5,
    color="skyblue",
    edgecolor="black",
    alpha=0.8
)

plt.title("Student Marks Distribution")

plt.xlabel("Marks")

plt.ylabel("Number of Students")

plt.grid(alpha=0.3)

plt.show()
```
