```python
# scatter() → Scatter Plot
# color → Single color
# c → Multiple colors
# cmap → Color theme
# s → Marker size
# marker → Marker shape
# alpha → Transparency
# colorbar() → Color scale
```

**1. What is Scatter Plot?**

```python
# Scatter Plot individual data points ko graph par show karta hai.
# Har point ek observation ko represent karta hai.
# Iska main purpose hai:
# Relationship dekhna
# Correlation samajhna
# Outliers identify karna
# Data distribution dekhna
```

**2. Why Scatter Plot?**

```python
# Use Cases

# Machine Learning
# Correlation Analysis
# Salary vs Experience
# Height vs Weight
# Temperature vs Ice Cream Sales
# Advertisement Cost vs Profit
```

**3. Basic Scatter Plot**

```python
import matplotlib.pyplot as plt

study_hours = [2,3,4,5,6,7,8]  # X-axis values.

marks = [40,50,60,65,75,85,95]  # Y-axis values.

plt.scatter(study_hours, marks)  # Scatter Plot draw karta hai.

plt.show()
```

**4. Change Marker Color**

```python
plt.scatter(
    study_hours,
    marks,
    color="red"
)

plt.show()
```

**5. Marker Size**

```python
# Default size small hota hai.
plt.scatter(
    study_hours,
    marks,
    s=200
)

plt.show()
```

**6. Marker Shape**

```python
plt.scatter(
    study_hours,
    marks,
    marker="s"
)

plt.show()

# Common Markers
# o → Circle

# s → Square

# ^ → Triangle

# * → Star

# D → Diamond

# x → Cross

# + → Plus
```

**7. Alpha (Transparency)**

```python
# 0 = Invisible
# 1 = Fully Visible

plt.scatter(
    study_hours,
    marks,
    alpha=0.5
)

plt.show()
```

**8. Color Mapping**

```python
colors = [10,20,30,40,50,60,70]

plt.scatter(
    study_hours,
    marks,
    c=colors,
    cmap="viridis"
)

plt.show()

# Popular Colormaps

# viridis

# plasma

# cool

# hot

# rainbow

# winter

# summer
```

**9. Color Bar**

```python
colors = [10,20,30,40,50,60,70]

scatter = plt.scatter(
    study_hours,
    marks,
    c=colors,
    cmap="viridis"
)
# Color Bar batata hai
# Kaunsa color kis value ko represent karta hai.
plt.colorbar(scatter)

plt.show()
```

**10. Combined Example**

```python
import matplotlib.pyplot as plt

study_hours = [2,3,4,5,6,7,8]

marks = [40,50,60,65,75,85,95]

colors = [20,30,40,50,60,70,80]

plt.figure(figsize=(8,5))

scatter = plt.scatter(
    study_hours,
    marks,
    c=colors,
    cmap="viridis",
    s=200,
    marker="o",
    alpha=0.8
)

plt.colorbar(scatter)

plt.title("Study Hours vs Marks")

plt.xlabel("Study Hours")

plt.ylabel("Marks")

plt.grid(alpha=0.3)

plt.show()
```
