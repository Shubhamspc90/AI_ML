```python
# Notes
# figure() → Figure create karta hai.
# figsize → Width & Height.
# dpi → Image quality.
# xlim() → X-axis range.
# ylim() → Y-axis range.
# xticks() → X labels customize.
# yticks() → Y labels customize.
# rotation → Labels rotate.
# fontsize → Font size.
# fontweight → Bold/Normal.
# savefig() → Graph save.
# style.use() → Graph theme.
# tight_layout() → Automatic spacing.
```

**1. Figure Size**

```python
# Figure Size graph ki width aur height decide karti hai.
# Default graph ka size fixed hota hai.
# Professional reports me hum khud size decide karte hain.
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May"]

sales = [50,70,65,90,100]

plt.figure(figsize=(8,5))

plt.plot(months,sales)

plt.show()
```

**2. DPI (Dots Per Inch)**

```python
# DPI image ki quality batata hai.
plt.figure(figsize=(8,5), dpi=150)
plt.plot(months,sales)
plt .show()
```

**3. xlim()**

```python
# X-axis ki range set karta hai.
plt.plot(months,sales)

plt.xlim(0,4)

plt.show()
```

**4. ylim()**

```python
# Y-axis ki range set karta hai.
plt.plot(months,sales)
plt.ylim(40,120)
plt.show()
```

**5. xticks()**

```python
# X-axis ke tick labels ko customize karta hai.
plt.plot(months,sales)

plt.xticks(fontsize=12)

plt.show()
```

**6. yticks()**

```python
plt.plot(months,sales)

plt.yticks(fontsize=12)

plt.show()
```

```python
# Legend automatically dono lines ko identify karega.
months=["Jan","Feb","Mar","Apr","May"]
product_A=[50,70,65,80,100]
product_B=[40,60,55,75,95]

plt.plot(months,product_A,label="Product_A")
plt.plot(months,product_B,label="Product_B")
plt.legend()
plt.show()
```

**15. Combined Professional Example**

```python
# Matplotlib library ko import kar rahe hain plotting ke liye
import matplotlib.pyplot as plt

# X-axis ke liye months ki list
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

# Y-axis ke liye sales data
sales = [50, 70, 65, 90, 100, 110]

# Graph ka overall style set kar rahe hain
plt.style.use("ggplot")

# Figure ka size aur image quality (resolution) set kar rahe hain
plt.figure(figsize=(10, 6), dpi=150)

# Line graph draw kar rahe hain
plt.plot(

    # X-axis ka data
    months,

    # Y-axis ka data
    sales,

    # Line ka color blue hoga
    color="blue",

    # Line style solid (-) hogi
    linestyle="-",

    # Line ki thickness 3 pixels hogi
    linewidth=3,

    # Har data point par circle marker lagega
    marker="o",

    # Marker ka size 8 hoga
    markersize=8,

    # Legend me is line ka naam "Sales" dikhai dega
    label="Sales"
)

# Graph ka title set kar rahe hain
plt.title(

    # Title ka text
    "Monthly Sales Report",

    # Title ka font size
    fontsize=18,

    # Title ko bold bana rahe hain
    fontweight="bold"
)

# X-axis ka label set kar rahe hain
plt.xlabel(

    # Label ka text
    "Months",

    # Label ka font size
    fontsize=14
)

# Y-axis ka label set kar rahe hain
plt.ylabel(

    # Label ka text
    "Sales",

    # Label ka font size
    fontsize=14
)

# X-axis ke labels ko 30 degree rotate kar rahe hain
plt.xticks(rotation=30)

# Grid show kar rahe hain aur uski transparency 30% rakh rahe hain
plt.grid(alpha=0.3)

# Legend show kar rahe hain
plt.legend()

# Graph ke elements ko automatically adjust karte hain
# Taaki labels aur title cut na ho
plt.tight_layout()

# Graph ko PNG image ke roop me save kar rahe hain
# dpi=300 ka matlab high quality image save hogi
plt.savefig("monthly_sales_report.png", dpi=300)

# Screen par graph display kar rahe hain
plt.show()
```
