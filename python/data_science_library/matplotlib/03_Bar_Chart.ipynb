```python
# Notes
# bar() → Vertical Bar Chart
# barh() → Horizontal Bar Chart
# color → Bar color
# width → Bar width
# xlabel() → X-axis label
# ylabel() → Y-axis label
# title() → Graph title
# grid() → Grid lines
# plt.text() → Value display
```

**1. What is Bar Chart?**

```python
# Theory

# Bar Chart ka use different categories ki values compare karne ke liye hota hai.
# Agar hume compare karna ho:
# Product Sales
# Student Marks
# Company Revenue
# Population
# Department Salary
# to Bar Chart best option hai.
```

**2. Basic Bar Chart**

```python
import matplotlib.pyplot as plt

products = ["Laptop", "Mouse", "Keyboard", "Monitor"]

sales = [25, 80, 60, 35]

plt.bar(products, sales)

plt.show()
```

**3. Change Bar Color**

```python
plt.bar(products, sales, color="green")

plt.show()
```

**4. Multiple Colors**

```python
# Har bar ka alag color.
plt.bar(
    products,
    sales,
    color=["red", "blue", "green", "orange"]
)

plt.show()
```

**5. Bar Width**

```python

# Default width hoti hai.,,Default width hoti hai.
plt.bar(
    products,
    sales,
    width=0.4
)

plt.show()
```

```python
# Thick bars
plt.bar(
    products,
    sales,
    width=0.8
)
plt.show()
```

**6. Horizontal Bar Chart**

```python
# Kab use hota hai?
# Jab category names bahut bade ho.

plt.barh(products, sales)

plt.show()
```

**7. Labels**

```python
plt.bar(products, sales)

plt.xlabel("Products")

plt.ylabel("Sales")

plt.show()
```

**8. Title**

```python
plt.bar(products,sales)
plt.title("Prodect Sales")
plt.show()
```

**9. Grid**

```python
plt.bar(products,sales)
# plt.grid(axis="y")
# plt.grid(axis="x")
plt.grid(axis="both")
plt.show()
```

**10. Display Values on Bars**

```python
products = ["Laptop", "Mouse", "Keyboard", "Monitor"]

sales = [25, 80, 60, 35]

bars = plt.bar(products, sales)

for bar in bars:

    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        bar.get_height(),
        ha="center"
    )

plt.show()

# Explanation
# bars me har bar ka object store hota hai.
# Loop har bar par chalta hai.
# plt.text() value ko bar ke upar print karta hai.
```

**11. Combined Example**

```python
import matplotlib.pyplot as plt

products = ["Laptop", "Mouse", "Keyboard", "Monitor"]

sales = [25, 80, 60, 35]

bars = plt.bar(
    products,
    sales,
    color=["red", "blue", "green", "orange"],
    width=0.6
)

plt.title("Monthly Product Sales")

plt.xlabel("Products")

plt.ylabel("Units Sold")

plt.grid(axis="y", alpha=0.3)

for bar in bars:

    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        bar.get_height(),
        ha="center"
    )

plt.show()
```
