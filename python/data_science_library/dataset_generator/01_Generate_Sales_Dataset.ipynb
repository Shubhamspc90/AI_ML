**💻 Cell 1 — Import Libraries + Configuration + Initial Setup**

```python
# ============================================================
# Sales Dataset Generator
# Author : Shubham Chauhan
# Project: Data Science Library
# ============================================================

# ===============================
# Import Libraries
# ===============================

import random
import numpy as np
import pandas as pd

from faker import Faker
from datetime import datetime, timedelta

# ===============================
# Create Faker Object
# ===============================

fake = Faker("en_IN")

# ===============================
# Random Seed
# ===============================

random.seed(42)
np.random.seed(42)

# ===============================
# Project Configuration
# ===============================

TOTAL_RECORDS = 10000

START_DATE = datetime(2022, 1, 1)
END_DATE   = datetime(2025, 12, 31)

EXPORT_CSV = True
EXPORT_EXCEL = True

OUTPUT_FOLDER = "../datasets/raw/"

# ===============================
# Display Settings
# ===============================

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 20)
pd.set_option("display.width", None)

# ===============================
# Project Information
# ===============================

print("=" * 60)
print("Sales Dataset Generator Started")
print("=" * 60)

print(f"Total Records : {TOTAL_RECORDS}")
print(f"Start Date    : {START_DATE.date()}")
print(f"End Date      : {END_DATE.date()}")

print("=" * 60)
print("Setup Completed Successfully")
print("=" * 60)
```

**💻 Code Cell 2**

```python
# ============================================================
# Master Lookup Tables
# ============================================================

# ----------------------------
# Product Categories
# ----------------------------

products = {
    "Electronics": [
        "Laptop", "Smartphone", "Tablet", "Smartwatch",
        "Headphones", "Bluetooth Speaker", "Power Bank",
        "Monitor", "Keyboard", "Mouse"
    ],

    "Furniture": [
        "Office Chair", "Dining Table", "Sofa", "Bed",
        "Wardrobe", "Bookshelf", "Study Table",
        "TV Unit", "Coffee Table", "Recliner"
    ],

    "Clothing": [
        "T-Shirt", "Jeans", "Jacket", "Shirt",
        "Hoodie", "Sneakers", "Kurta",
        "Track Pant", "Sweater", "Blazer"
    ],

    "Home & Kitchen": [
        "Mixer", "Microwave", "Pressure Cooker",
        "Water Bottle", "Ceiling Fan",
        "Gas Stove", "Cookware Set",
        "Vacuum Cleaner", "Iron", "Air Fryer"
    ],

    "Sports": [
        "Cricket Bat", "Football",
        "Basketball", "Badminton Racket",
        "Gym Gloves", "Yoga Mat",
        "Skipping Rope", "Tennis Ball",
        "Dumbbell", "Helmet"
    ]
}

# ----------------------------
# Brands
# ----------------------------

brands = [
    "Samsung",
    "Apple",
    "Sony",
    "HP",
    "Dell",
    "Lenovo",
    "Boat",
    "LG",
    "Nike",
    "Adidas",
    "Puma",
    "Reebok",
    "Prestige",
    "Philips",
    "Bajaj",
    "Godrej",
    "Nilkamal",
    "Cello",
    "Milton",
    "Havells"
]

# ----------------------------
# Payment Modes
# ----------------------------

payment_modes = [
    "UPI",
    "Cash",
    "Debit Card",
    "Credit Card",
    "Net Banking"
]

# ----------------------------
# Order Priority
# ----------------------------

order_priority = [
    "Low",
    "Medium",
    "High",
    "Critical"
]

# ----------------------------
# Customer Segment
# ----------------------------

customer_segment = [
    "Regular",
    "Premium",
    "Corporate"
]

print("="*60)
print("Master Lookup Tables Created Successfully")
print("="*60)

print(f"Categories       : {len(products)}")
print(f"Brands           : {len(brands)}")
print(f"Payment Modes    : {len(payment_modes)}")
print(f"Order Priorities : {len(order_priority)}")
print(f"Customer Segments: {len(customer_segment)}")

print("="*60)
```

**💻 Code Cell 4**

```python
# ============================================================
# Create Product Master
# ============================================================

product_master = []

product_id = 1001

for category, product_list in products.items():

    for product in product_list:

        for brand in random.sample(brands, 3):

            price = random.randint(500, 100000)

            cost = round(price * random.uniform(0.55, 0.85), 2)

            product_master.append({

                "Product_ID": f"P{product_id}",

                "Product_Name": product,

                "Category": category,

                "Brand": brand,

                "Unit_Price": price,

                "Cost": cost

            })

            product_id += 1

# ============================================================
# Convert into DataFrame
# ============================================================

product_df = pd.DataFrame(product_master)

print("=" * 60)
print("Product Master Created Successfully")
print("=" * 60)

print(f"Total Products : {len(product_df)}")

print()

print(product_df.head())

print("=" * 60)
```

**💻 Code Cell 5**

```python
# ============================================================
# Create Customer Master
# ============================================================

customer_master = []

TOTAL_CUSTOMERS = 4000

for customer_no in range(1, TOTAL_CUSTOMERS + 1):

    # -------------------------
    # Select Region
    # -------------------------

    region = random.choice(list(india_locations.keys()))

    # -------------------------
    # Select State
    # -------------------------

    state = random.choice(list(india_locations[region].keys()))

    # -------------------------
    # Select City
    # -------------------------

    city = random.choice(india_locations[region][state])

    # -------------------------
    # Gender
    # -------------------------

    gender = random.choice(["Male", "Female"])

    # -------------------------
    # Name
    # -------------------------

    if gender == "Male":
        customer_name = fake.name_male()
    else:
        customer_name = fake.name_female()

    # -------------------------
    # Customer Record
    # -------------------------

    customer_master.append({

        "Customer_ID": f"CUST{customer_no:05d}",

        "Customer_Name": customer_name,

        "Gender": gender,

        "Age": random.randint(18, 70),

        "Email": fake.email(),

        "Phone": fake.phone_number(),

        "Region": region,

        "State": state,

        "City": city

    })

# ============================================================
# Convert into DataFrame
# ============================================================

customer_df = pd.DataFrame(customer_master)

print("="*60)
print("Customer Master Created Successfully")
print("="*60)

print(f"Total Customers : {len(customer_df)}")

print()

print(customer_df.head())

print("="*60)
```

**💻 Code Cell 6**

```python
# ============================================================
# Generate Sales Orders
# ============================================================

sales_data = []

# Total Days Between Start & End Date
date_range = (END_DATE - START_DATE).days

for order_no in range(1, TOTAL_RECORDS + 1):

    # ------------------------------------------------
    # Order ID
    # ------------------------------------------------

    order_id = f"ORD{order_no:06d}"

    # ------------------------------------------------
    # Random Order Date
    # ------------------------------------------------

    order_date = START_DATE + timedelta(
        days=random.randint(0, date_range)
    )

    # ------------------------------------------------
    # Shipping Date
    # ------------------------------------------------

    delivery_days = random.randint(1, 7)

    ship_date = order_date + timedelta(days=delivery_days)

    # ------------------------------------------------
    # Random Customer
    # ------------------------------------------------

    customer = customer_df.sample(1).iloc[0]

    # ------------------------------------------------
    # Random Product
    # ------------------------------------------------

    product = product_df.sample(1).iloc[0]

    # ------------------------------------------------
    # Quantity
    # ------------------------------------------------

    quantity = random.randint(1, 10)

    # ------------------------------------------------
    # Discount
    # ------------------------------------------------

    discount = random.choice([
        0,
        5,
        10,
        15,
        20,
        25,
        30
    ])

    # ------------------------------------------------
    # Sales
    # ------------------------------------------------

    sales = product["Unit_Price"] * quantity

    sales_after_discount = sales - (sales * discount / 100)

    # ------------------------------------------------
    # Cost
    # ------------------------------------------------

    total_cost = product["Cost"] * quantity

    # ------------------------------------------------
    # Profit
    # ------------------------------------------------

    profit = sales_after_discount - total_cost

    # ------------------------------------------------
    # Returned
    # ------------------------------------------------

    returned = random.choice([
        "Yes",
        "No",
        "No",
        "No",
        "No"
    ])

    # ------------------------------------------------
    # Customer Rating
    # ------------------------------------------------

    rating = random.randint(1, 5)

    # ------------------------------------------------
    # Save Record
    # ------------------------------------------------

    sales_data.append({

        "Order_ID": order_id,

        "Order_Date": order_date.date(),

        "Ship_Date": ship_date.date(),

        "Customer_ID": customer["Customer_ID"],

        "Customer_Name": customer["Customer_Name"],

        "Gender": customer["Gender"],

        "Age": customer["Age"],

        "Email": customer["Email"],

        "Phone": customer["Phone"],

        "Region": customer["Region"],

        "State": customer["State"],

        "City": customer["City"],

        "Product_ID": product["Product_ID"],

        "Product_Name": product["Product_Name"],

        "Category": product["Category"],

        "Brand": product["Brand"],

        "Quantity": quantity,

        "Unit_Price": product["Unit_Price"],

        "Discount": discount,

        "Sales": round(sales_after_discount, 2),

        "Cost": round(total_cost, 2),

        "Profit": round(profit, 2),

        "Payment_Mode": random.choice(payment_modes),

        "Order_Priority": random.choice(order_priority),

        "Customer_Segment": random.choice(customer_segment),

        "Delivery_Days": delivery_days,

        "Customer_Rating": rating,

        "Returned": returned

    })

print("=" * 60)
print("Sales Orders Generated Successfully")
print("=" * 60)
print(f"Total Orders : {len(sales_data)}")
print("=" * 60)
```

**💻 Code Cell 7**

```python
# ============================================================
# Convert to DataFrame
# ============================================================

sales_df = pd.DataFrame(sales_data)

print("=" * 60)
print("Sales DataFrame Created")
print("=" * 60)

print()

print(sales_df.head())

print()

print(sales_df.shape)
```

**💻 Code Cell 8**

```python
# ============================================================
# Add Missing Values
# ============================================================

missing_columns = [
    "Email",
    "Phone",
    "Customer_Rating",
    "Discount"
]

for column in missing_columns:

    random_index = sales_df.sample(frac=0.02, random_state=42).index

    sales_df.loc[random_index, column] = np.nan

print("=" * 60)
print("Missing Values Added Successfully")
print("=" * 60)

print(sales_df.isnull().sum())

print("=" * 60)
```

**💻 Code Cell 9**

```python
# ============================================================
# Add Duplicate Records
# ============================================================

duplicate_rows = sales_df.sample(
    n=100,
    random_state=42
)

sales_df = pd.concat(
    [sales_df, duplicate_rows],
    ignore_index=True
)

print("=" * 60)
print("Duplicate Records Added Successfully")
print("=" * 60)

print("Total Rows :", len(sales_df))

print("=" * 60)
```

**💻 Code Cell 10**

```python
# ============================================================
# Add Outliers
# ============================================================

outlier_index = sales_df.sample(
    n=50,
    random_state=100
).index

sales_df.loc[outlier_index, "Sales"] *= 8

sales_df.loc[outlier_index, "Profit"] *= 10

print("=" * 60)
print("Outliers Added Successfully")
print("=" * 60)
```

**💻 Code Cell 11**

```python
# ============================================================
# Add Invalid Values
# ============================================================

invalid_index = sales_df.sample(
    n=30,
    random_state=200
).index

sales_df.loc[invalid_index, "Age"] = -1

print("=" * 60)
print("Invalid Values Added Successfully")
print("=" * 60)
```
