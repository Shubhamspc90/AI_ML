**💻 Code Cell 5**

```python
# ============================================================
# Create Customer Master
# ============================================================

customer_master = []

TOTAL_CUSTOMERS = 4000

for customer_no in range(1, TOTAL_CUSTOMERS + 1):

    # -------------------------
    # Select Region
    # -------------------------

    region = random.choice(list(india_locations.keys()))

    # -------------------------
    # Select State
    # -------------------------

    state = random.choice(list(india_locations[region].keys()))

    # -------------------------
    # Select City
    # -------------------------

    city = random.choice(india_locations[region][state])

    # -------------------------
    # Gender
    # -------------------------

    gender = random.choice(["Male", "Female"])

    # -------------------------
    # Name
    # -------------------------

    if gender == "Male":
        customer_name = fake.name_male()
    else:
        customer_name = fake.name_female()

    # -------------------------
    # Customer Record
    # -------------------------

    customer_master.append({

        "Customer_ID": f"CUST{customer_no:05d}",

        "Customer_Name": customer_name,

        "Gender": gender,

        "Age": random.randint(18, 70),

        "Email": fake.email(),

        "Phone": fake.phone_number(),

        "Region": region,

        "State": state,

        "City": city

    })

# ============================================================
# Convert into DataFrame
# ============================================================

customer_df = pd.DataFrame(customer_master)

print("="*60)
print("Customer Master Created Successfully")
print("="*60)

print(f"Total Customers : {len(customer_df)}")

print()

print(customer_df.head())

print("="*60)
```

**💻 Code Cell 6**

```python
# ============================================================
# Generate Sales Orders
# ============================================================

sales_data = []

# Total Days Between Start & End Date
date_range = (END_DATE - START_DATE).days

for order_no in range(1, TOTAL_RECORDS + 1):

    # ------------------------------------------------
    # Order ID
    # ------------------------------------------------

    order_id = f"ORD{order_no:06d}"

    # ------------------------------------------------
    # Random Order Date
    # ------------------------------------------------

    order_date = START_DATE + timedelta(
        days=random.randint(0, date_range)
    )

    # ------------------------------------------------
    # Shipping Date
    # ------------------------------------------------

    delivery_days = random.randint(1, 7)

    ship_date = order_date + timedelta(days=delivery_days)

    # ------------------------------------------------
    # Random Customer
    # ------------------------------------------------

    customer = customer_df.sample(1).iloc[0]

    # ------------------------------------------------
    # Random Product
    # ------------------------------------------------

    product = product_df.sample(1).iloc[0]

    # ------------------------------------------------
    # Quantity
    # ------------------------------------------------

    quantity = random.randint(1, 10)

    # ------------------------------------------------
    # Discount
    # ------------------------------------------------

    discount = random.choice([
        0,
        5,
        10,
        15,
        20,
        25,
        30
    ])

    # ------------------------------------------------
    # Sales
    # ------------------------------------------------

    sales = product["Unit_Price"] * quantity

    sales_after_discount = sales - (sales * discount / 100)

    # ------------------------------------------------
    # Cost
    # ------------------------------------------------

    total_cost = product["Cost"] * quantity

    # ------------------------------------------------
    # Profit
    # ------------------------------------------------

    profit = sales_after_discount - total_cost

    # ------------------------------------------------
    # Returned
    # ------------------------------------------------

    returned = random.choice([
        "Yes",
        "No",
        "No",
        "No",
        "No"
    ])

    # ------------------------------------------------
    # Customer Rating
    # ------------------------------------------------

    rating = random.randint(1, 5)

    # ------------------------------------------------
    # Save Record
    # ------------------------------------------------

    sales_data.append({

        "Order_ID": order_id,

        "Order_Date": order_date.date(),

        "Ship_Date": ship_date.date(),

        "Customer_ID": customer["Customer_ID"],

        "Customer_Name": customer["Customer_Name"],

        "Gender": customer["Gender"],

        "Age": customer["Age"],

        "Email": customer["Email"],

        "Phone": customer["Phone"],

        "Region": customer["Region"],

        "State": customer["State"],

        "City": customer["City"],

        "Product_ID": product["Product_ID"],

        "Product_Name": product["Product_Name"],

        "Category": product["Category"],

        "Brand": product["Brand"],

        "Quantity": quantity,

        "Unit_Price": product["Unit_Price"],

        "Discount": discount,

        "Sales": round(sales_after_discount, 2),

        "Cost": round(total_cost, 2),

        "Profit": round(profit, 2),

        "Payment_Mode": random.choice(payment_modes),

        "Order_Priority": random.choice(order_priority),

        "Customer_Segment": random.choice(customer_segment),

        "Delivery_Days": delivery_days,

        "Customer_Rating": rating,

        "Returned": returned

    })

print("=" * 60)
print("Sales Orders Generated Successfully")
print("=" * 60)
print(f"Total Orders : {len(sales_data)}")
print("=" * 60)
```

**💻 Code Cell 17**

```python
# ============================================================
# Helper Functions
# ============================================================

def random_order_date():
    """Generate random order date"""

    days = random.randint(
        0,
        (END_DATE - START_DATE).days
    )

    return START_DATE + timedelta(days=days)


def random_discount(segment):

    if segment == "Corporate":
        return random.choice([5,10,15,20,25,30])

    elif segment == "Premium":
        return random.choice([0,5,10])

    else:
        return random.choice([0,5,10,15,20])


def random_rating(returned):

    if returned == "Yes":

        return random.choice([1,2])

    else:

        return random.choice([3,4,5])


def calculate_profit(sales,cost):

    return round(sales-cost,2)


print("="*60)
print("Helper Functions Ready")
print("="*60)
```

**💻 Code Cell 21**

```python
# ============================================================
# Quantity Generator
# ============================================================

def generate_quantity(segment):

    if segment == "Corporate":

        return random.randint(5,20)

    elif segment == "Premium":

        return random.randint(1,3)

    else:

        return random.randint(1,5)


print(generate_quantity("Regular"))
print(generate_quantity("Premium"))
print(generate_quantity("Corporate"))
```

**💻 Code Cell 22**

```python
# ============================================================
# Sales Calculator
# ============================================================

def calculate_sales(price, quantity, discount):

    gross = price * quantity

    discount_amount = gross * (discount / 100)

    final_sales = gross - discount_amount

    return round(final_sales,2)


print(calculate_sales(5000,3,10))
```

**💻 Code Cell 23**

```python
# ============================================================
# Cost Calculator
# ============================================================

def calculate_cost(cost_price, quantity):

    return round(cost_price * quantity,2)


print(calculate_cost(2500,3))
```

**💻 Code Cell 24**

```python
# ============================================================
# Return Logic
# ============================================================

def get_return_status(rating):

    if rating <= 2:

        return "Yes"

    return random.choice([
        "No",
        "No",
        "No",
        "No",
        "Yes"
    ])


print(get_return_status(5))
print(get_return_status(1))
```

**💻 Code Cell 25**

```python
# ============================================================
# Shipping Days
# ============================================================

def get_delivery_days(region):

    if region == "North":

        return random.randint(1,3)

    elif region == "West":

        return random.randint(2,4)

    elif region == "South":

        return random.randint(3,5)

    else:

        return random.randint(4,7)


print(get_delivery_days("North"))
print(get_delivery_days("South"))
```

**💻 Code Cell 26**

```python
# ============================================================
# Shipping Date
# ============================================================

def get_ship_date(order_date, delivery_days):

    return order_date + timedelta(days=delivery_days)
```

**💻 Code Cell 27**

```python
# ============================================================
# Profit Calculator
# ============================================================

def calculate_profit(sales, cost):

    return round(sales - cost,2)


print(calculate_profit(50000,42000))
```

**💻 Code Cell 28**

```python
# ============================================================
# Generate Sales Orders
# ============================================================

def generate_orders(total_orders):

    sales_data = []

    for _ in range(total_orders):

        # ------------------------------------------------
        # Order ID
        # ------------------------------------------------

        order_id = generate_order_id()

        # ------------------------------------------------
        # Customer
        # ------------------------------------------------

        customer = get_customer()

        # ------------------------------------------------
        # Product
        # ------------------------------------------------

        product = get_product()

        # ------------------------------------------------
        # Customer Segment
        # ------------------------------------------------

        segment = random.choice(customer_segment)

        # ------------------------------------------------
        # Order Date
        # ------------------------------------------------

        order_date = random_order_date()

        # ------------------------------------------------
        # Quantity
        # ------------------------------------------------

        quantity = generate_quantity(segment)

        # ------------------------------------------------
        # Discount
        # ------------------------------------------------

        discount = random_discount(segment)

        # ------------------------------------------------
        # Sales
        # ------------------------------------------------

        sales = calculate_sales(
            product["Unit_Price"],
            quantity,
            discount
        )

        # ------------------------------------------------
        # Cost
        # ------------------------------------------------

        cost = calculate_cost(
            product["Cost"],
            quantity
        )

        # ------------------------------------------------
        # Profit
        # ------------------------------------------------

        profit = calculate_profit(
            sales,
            cost
        )

        # ------------------------------------------------
        # Rating
        # ------------------------------------------------

        rating = random.randint(1, 5)

        # ------------------------------------------------
        # Returned
        # ------------------------------------------------

        returned = get_return_status(rating)

        # ------------------------------------------------
        # Delivery
        # ------------------------------------------------

        delivery_days = get_delivery_days(
            customer["Region"]
        )

        ship_date = get_ship_date(
            order_date,
            delivery_days
        )

        # ------------------------------------------------
        # Save Record
        # ------------------------------------------------

        sales_data.append({

            "Order_ID": order_id,

            "Order_Date": order_date.date(),

            "Ship_Date": ship_date.date(),

            "Customer_ID": customer["Customer_ID"],

            "Customer_Name": customer["Customer_Name"],

            "Gender": customer["Gender"],

            "Age": customer["Age"],

            "Email": customer["Email"],

            "Phone": customer["Phone"],

            "Region": customer["Region"],

            "State": customer["State"],

            "City": customer["City"],

            "Product_ID": product["Product_ID"],

            "Product_Name": product["Product_Name"],

            "Category": product["Category"],

            "Brand": product["Brand"],

            "Quantity": quantity,

            "Unit_Price": product["Unit_Price"],

            "Discount": discount,

            "Sales": sales,

            "Cost": cost,

            "Profit": profit,

            "Payment_Mode": random.choice(payment_modes),

            "Order_Priority": random.choice(order_priority),

            "Customer_Segment": segment,

            "Delivery_Days": delivery_days,

            "Customer_Rating": rating,

            "Returned": returned

        })

    return pd.DataFrame(sales_data)
```
